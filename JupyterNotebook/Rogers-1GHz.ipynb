## üìÇ SECCI√ìN 2: Carga y Procesamiento de Datos S-Parameter

Cargamos los datos del archivo Rogers CSV y procesamos los par√°metros S11 y S21.

````python
# ============================================================
# üìÇ CARGA Y AN√ÅLISIS DE DATOS S21 - ROGERS 1GHz
# ============================================================

# CONFIGURACI√ìN ESPEC√çFICA PARA ROGERS-1GHz.s2p
ARCHIVO_S2P = 'Rogers-1GHz.s2p'

def cargar_datos_rogers_s2p():
    """
    Carga datos S21 espec√≠ficamente del archivo Rogers-1GHz.s2p
    Formato Touchstone: Freq(GHz) S11_mag S11_ang S12_mag S12_ang S21_mag S21_ang S22_mag S22_ang
    Salida: frecuencia (GHz), S21_dB, S21_complex
    """
    if not os.path.exists(ARCHIVO_S2P):
        print(f"‚ùå Archivo no encontrado: {ARCHIVO_S2P}")
        return None
    
    try:
        # Leer el archivo s2p saltando las l√≠neas de comentario
        data = []
        with open(ARCHIVO_S2P, 'r') as file:
            for line in file:
                line = line.strip()
                # Saltar l√≠neas de comentario y encabezados
                if line.startswith('!') or line.startswith('#') or not line:
                    continue
                # Convertir l√≠nea a n√∫meros
                values = [float(x) for x in line.split()]
                if len(values) == 9:  # Formato correcto: freq + 8 par√°metros S
                    data.append(values)
        
        # Convertir a array numpy
        data = np.array(data)
        
        # Extraer columnas seg√∫n formato Touchstone
        # Columnas: 0=freq, 1=S11_mag, 2=S11_ang, 3=S12_mag, 4=S12_ang, 5=S21_mag, 6=S21_ang, 7=S22_mag, 8=S22_ang
        freq_ghz = data[:, 0]           # Frecuencia en GHz
        s21_mag = data[:, 5]            # Magnitud S21
        s21_ang_deg = data[:, 6]        # √Ångulo S21 en grados
        
        # Convertir a forma compleja
        s21_ang_rad = np.deg2rad(s21_ang_deg)  # Convertir a radianes
        s21_complex = s21_mag * np.exp(1j * s21_ang_rad)
        
        # Convertir magnitud a dB
        s21_db = 20 * np.log10(s21_mag)
        
        return freq_ghz, s21_db, s21_complex
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error al procesar {ARCHIVO_S2P}: {e}")
        return None

# ============================================================
# EJECUCI√ìN Y VARIABLES PARA SECCIONES SIGUIENTES
# ============================================================

# Cargar datos espec√≠ficos del archivo Rogers
resultado = cargar_datos_rogers_s2p()

if resultado is not None:
    # Variables principales para todo el notebook
    frecuencias = resultado[0]     # Array frecuencias en GHz
    s21_magnitud = resultado[1]    # Array S21 en dB  
    s21_complejo = resultado[2]    # Array S21 complejo
    
    # M√©tricas b√°sicas
    n_puntos = len(frecuencias)
    f_min, f_max = frecuencias[0], frecuencias[-1] 
    s21_min, s21_max = s21_magnitud.min(), s21_magnitud.max()
    s21_medio = np.mean(s21_magnitud)
    rango_dinamico = s21_max - s21_min
    
    # Verificar integridad
    datos_validos = not (np.any(np.isnan(s21_magnitud)) or np.any(np.isinf(s21_magnitud)))
    
    # Resumen consolidado (una sola salida)
    print(f"‚úÖ ROGERS-1GHz.s2p procesado correctamente:")
    print(f"   üìä {n_puntos:,} puntos | {f_min:.1f}-{f_max:.1f} GHz")
    print(f"   üìà S21: {s21_min:.1f} a {s21_max:.1f} dB (promedio: {s21_medio:.1f} dB)")
    print(f"   üéØ Rango din√°mico: {rango_dinamico:.1f} dB | {'‚úì' if datos_validos else '‚ö†Ô∏è'} Datos {'v√°lidos' if datos_validos else 'con errores'}")
        
else:
    print("‚ùå No se pudieron cargar los datos del archivo Rogers-1GHz.s2p")
````

Encontramos autom√°ticamente las frecuencias donde el resonador de anillo presenta picos de transmisi√≥n.

````python
# ============================================================
# üéØ DETECCI√ìN SIMPLE DE RESONANCIAS
# ============================================================

def detectar_picos_simples(frecuencias, s21_db, altura_minima=-50, separacion_minima=50):
    """
    Detecta picos de resonancia de manera simple y directa
    
    Par√°metros:
    - altura_minima: Los picos deben estar por encima de este nivel (dB)
    - separacion_minima: M√≠nima separaci√≥n entre picos (puntos de datos)
    """
    
    # Usar scipy para encontrar picos (mantenemos prominencia interna para filtrar correctamente)
    indices_picos, propiedades = signal.find_peaks(
        s21_db,
        height=altura_minima,
        prominence=3,  # Mantener prominencia interna para detectar solo picos reales
        distance=separacion_minima
    )
    
    # Extraer informaci√≥n b√°sica de cada pico (sin mostrar prominencia)
    picos_encontrados = []
    
    for i, indice in enumerate(indices_picos):
        pico = {
            'numero': i + 1,
            'indice': indice,
            'frecuencia_ghz': frecuencias[indice],
            'frecuencia_mhz': frecuencias[indice] * 1000,
            's21_db': s21_db[indice]
        }
        picos_encontrados.append(pico)
    
    return picos_encontrados, indices_picos

# ============================================================
# EJECUTAR DETECCI√ìN
# ============================================================

print("üîç Buscando picos de resonancia...")

# Detectar picos con par√°metros ajustables
picos_detectados, indices_picos = detectar_picos_simples(
    frecuencias, 
    s21_magnitud, 
    altura_minima=-50,      # Solo picos por encima de -50 dB
    separacion_minima=100   # Separaci√≥n m√≠nima entre picos
)

print(f"\n‚úÖ ENCONTRADAS {len(picos_detectados)} RESONANCIAS:")
print("-" * 60)

for pico in picos_detectados:
    print(f"üéØ Resonancia {pico['numero']}: {pico['frecuencia_ghz']:.4f} GHz ({pico['frecuencia_mhz']:.1f} MHz)")
    print(f"   ‚îÇ Nivel S21: {pico['s21_db']:.1f} dB")
    print()

# ============================================================
# VISUALIZACI√ìN SIMPLE
# ============================================================

plt.figure(figsize=(12, 6))

# Gr√°fico base de S21
plt.plot(frecuencias, s21_magnitud, color=COLORES['datos'], linewidth=1.5, 
         label='S21 medido', alpha=0.8)

# Marcar los picos encontrados
if len(picos_detectados) > 0:
    frecuencias_picos = [p['frecuencia_ghz'] for p in picos_detectados]
    niveles_picos = [p['s21_db'] for p in picos_detectados]
    
    plt.scatter(frecuencias_picos, niveles_picos, 
                color=COLORES['resonancia'], s=80, zorder=5,
                label=f'{len(picos_detectados)} resonancias detectadas')
    
    # Etiquetar cada pico
    for pico in picos_detectados:
        plt.annotate(f'R{pico["numero"]}\n{pico["frecuencia_ghz"]:.3f} GHz', 
                    xy=(pico['frecuencia_ghz'], pico['s21_db']),
                    xytext=(10, 15), textcoords='offset points',
                    fontsize=9, ha='center',
                    bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.8),
                    arrowprops=dict(arrowstyle='->', color='black', alpha=0.7))

plt.xlabel('Frecuencia (GHz)')
plt.ylabel('|S21| (dB)')
plt.title(' ')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Guardar resultados para siguientes secciones
resonancias_detectadas = picos_detectados
n_resonancias = len(picos_detectados)

print(f"‚úÖ Detecci√≥n completada - {n_resonancias} resonancias listas para an√°lisis detallado")
````

## üìä SECCI√ìN 4: Factor Q y Ancho de Banda

Calculamos los par√°metros fundamentales de cada resonancia: **factor Q**, **ancho de banda** y **frecuencia de resonancia**.

````python
# ============================================================
# üìä AN√ÅLISIS DE FACTOR Q Y ANCHO DE BANDA
# ============================================================

def calcular_factor_q(frecuencias, s21_db, indice_pico):
    """
    Calcula el factor de calidad Q de una resonancia usando el m√©todo -3dB
    
    Q = f‚ÇÄ / Œîf‚ÇÉdB
    
    Donde:
    - f‚ÇÄ = frecuencia de resonancia
    - Œîf‚ÇÉdB = ancho de banda a -3dB del m√°ximo
    """
    
    freq_res = frecuencias[indice_pico]
    s21_max = s21_db[indice_pico]
    
    # Nivel de referencia a -3dB del m√°ximo
    nivel_3db = s21_max - 3
    
    # Buscar punto izquierdo (frecuencias menores)
    idx_izq = indice_pico
    while idx_izq > 0 and s21_db[idx_izq] >= nivel_3db:
        idx_izq -= 1
    
    # Buscar punto derecho (frecuencias mayores)
    idx_der = indice_pico
    while idx_der < len(s21_db)-1 and s21_db[idx_der] >= nivel_3db:
        idx_der += 1
    
    # Calcular ancho de banda y factor Q
    if idx_der > idx_izq:
        freq_izq = frecuencias[idx_izq]
        freq_der = frecuencias[idx_der]
        ancho_banda = freq_der - freq_izq
        
        factor_q = freq_res / ancho_banda if ancho_banda > 0 else 0
        
        return {
            'factor_q': factor_q,
            'ancho_banda_ghz': ancho_banda,
            'ancho_banda_mhz': ancho_banda * 1000,
            'freq_izq_ghz': freq_izq,
            'freq_der_ghz': freq_der,
            'valido': True
        }
    else:
        return {
            'factor_q': 0,
            'ancho_banda_ghz': 0,
            'ancho_banda_mhz': 0,
            'freq_izq_ghz': 0,
            'freq_der_ghz': 0,
            'valido': False
        }

# ============================================================
# AN√ÅLISIS DE CADA RESONANCIA
# ============================================================

print("üìä Calculando factor Q y ancho de banda...")
print("=" * 60)

resonancias_analizadas = []

for pico in resonancias_detectadas:
    # Calcular factor Q
    q_data = calcular_factor_q(frecuencias, s21_magnitud, pico['indice'])
    
    # Combinar informaci√≥n b√°sica con par√°metros Q
    resonancia_completa = {
        'numero': pico['numero'],
        'frecuencia_ghz': pico['frecuencia_ghz'],
        'frecuencia_mhz': pico['frecuencia_mhz'],
        's21_db': pico['s21_db'],  # Corregido: usar 's21_db' en lugar de 'nivel_db'
        'indice': pico['indice'],
        **q_data
    }
    
    resonancias_analizadas.append(resonancia_completa)
    
    # Mostrar resultados
    print(f"üéØ Resonancia {pico['numero']}: {pico['frecuencia_ghz']:.4f} GHz")
    print(f"   ‚îÇ Factor Q: {q_data['factor_q']:.1f}")
    print(f"   ‚îÇ Ancho de banda: {q_data['ancho_banda_mhz']:.2f} MHz")
    print(f"   ‚îÇ Nivel S‚ÇÇ‚ÇÅ: {pico['s21_db']:.1f} dB")
    print()

# ============================================================
# VISUALIZACI√ìN DE PAR√ÅMETROS
# ============================================================

if len(resonancias_analizadas) > 0:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # Gr√°fico 1: Factores Q
    factores_q = [r['factor_q'] for r in resonancias_analizadas if r['valido']]
    labels_q = [f"R{r['numero']}\n{r['frecuencia_ghz']:.3f}GHz" 
                for r in resonancias_analizadas if r['valido']]
    
    bars1 = ax1.bar(range(len(factores_q)), factores_q, 
                    color=COLORES['destacado'], alpha=0.7)
    ax1.set_xlabel('Resonancia')
    ax1.set_ylabel('Factor Q')
    ax1.set_title('Factor de Calidad Q')
    ax1.set_xticks(range(len(factores_q)))
    ax1.set_xticklabels(labels_q, rotation=45)
    ax1.grid(True, alpha=0.3)
    
    # Valores en las barras
    for i, bar in enumerate(bars1):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height + height*0.02,
                f'{factores_q[i]:.1f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    # Gr√°fico 2: Ancho de banda
    anchos_banda = [r['ancho_banda_mhz'] for r in resonancias_analizadas if r['valido']]
    
    bars2 = ax2.bar(range(len(anchos_banda)), anchos_banda, 
                    color=COLORES['resonancia'], alpha=0.7)
    ax2.set_xlabel('Resonancia')
    ax2.set_ylabel('Ancho de Banda (MHz)')
    ax2.set_title('Ancho de Banda a -3dB')
    ax2.set_xticks(range(len(anchos_banda)))
    ax2.set_xticklabels(labels_q, rotation=45)
    ax2.grid(True, alpha=0.3)
    
    # Valores en las barras
    for i, bar in enumerate(bars2):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + height*0.02,
                f'{anchos_banda[i]:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# ============================================================
# ESTAD√çSTICAS GENERALES
# ============================================================

resonancias_validas = [r for r in resonancias_analizadas if r['valido']]

if resonancias_validas:
    q_promedio = np.mean([r['factor_q'] for r in resonancias_validas])
    ancho_promedio = np.mean([r['ancho_banda_mhz'] for r in resonancias_validas])
    
    print(f"üìà ESTAD√çSTICAS GENERALES:")
    print(f"   ‚Ä¢ N√∫mero de resonancias: {len(resonancias_validas)}")
    print(f"   ‚Ä¢ Factor Q promedio: {q_promedio:.1f}")
    print(f"   ‚Ä¢ Ancho de banda promedio: {ancho_promedio:.2f} MHz")
    print(f"   ‚Ä¢ Rango de frecuencias: {resonancias_validas[0]['frecuencia_ghz']:.3f} - {resonancias_validas[-1]['frecuencia_ghz']:.3f} GHz")

print("\n‚úÖ An√°lisis de factor Q completado")
````

#### **üìê Fundamentos Matem√°ticos**

**Factor de Calidad Q:**
$$Q = \frac{f_0}{\Delta f_{3dB}}$$

**Par√°metros:**
- $f_0$ = frecuencia de resonancia (GHz)
- $\Delta f_{3dB}$ = ancho de banda a -3dB del m√°ximo (GHz)

**Ancho de Banda a -3dB:**

El ancho de banda se mide entre los puntos donde la potencia cae a la mitad del m√°ximo, equivalente a una reducci√≥n de 3dB en escala logar√≠tmica.

$$\Delta f_{3dB} = f_{der} - f_{izq}$$

Donde $f_{izq}$ y $f_{der}$ son las frecuencias donde $S_{21}(f) = S_{21,max} - 3dB$

**Interpretaci√≥n del Factor Q:**
- **Q alto**: Resonancia estrecha y bien definida
- **Q bajo**: Resonancia ancha con mayor p√©rdida de energ√≠a

## üéº SECCI√ìN 5: An√°lisis de Simetr√≠a entre Modos

Analizamos las **relaciones arm√≥nicas** entre las resonancias y evaluamos la **uniformidad** de los modos del resonador en anillo.

````python
# ============================================================
# üéº AN√ÅLISIS DE SIMETR√çA ENTRE MODOS - VERSI√ìN CONCISA
# ============================================================

def analizar_simetria_modos(resonancias_analizadas):
    """
    Analiza las relaciones arm√≥nicas y simetr√≠a entre modos de resonancia
    """
    
    # Extraer frecuencias ordenadas
    frecuencias = [r['frecuencia_ghz'] for r in resonancias_analizadas if r['valido']]
    frecuencias.sort()
    
    if len(frecuencias) < 2:
        return None
    
    # Frecuencia fundamental (la m√°s baja)
    f_fundamental = frecuencias[0]
    
    # Calcular relaciones arm√≥nicas
    relaciones_armonicas = []
    desviaciones = []
    
    for i, freq in enumerate(frecuencias):
        relacion_teorica = i + 1  # 1, 2, 3, ...
        relacion_real = freq / f_fundamental
        desviacion_porcentual = abs(relacion_real - relacion_teorica) / relacion_teorica * 100
        
        relaciones_armonicas.append({
            'modo': i + 1,
            'frecuencia_ghz': freq,
            'relacion_teorica': relacion_teorica,
            'relacion_real': relacion_real,
            'desviacion_porcentual': desviacion_porcentual
        })
        
        desviaciones.append(desviacion_porcentual)
    
    # Calcular intervalos entre modos consecutivos
    intervalos = []
    for i in range(1, len(frecuencias)):
        intervalo = frecuencias[i] - frecuencias[i-1]
        intervalos.append(intervalo)
    
    # Uniformidad de intervalos
    if intervalos:
        intervalo_promedio = np.mean(intervalos)
        desviacion_intervalos = np.std(intervalos)
        coeficiente_variacion = desviacion_intervalos / intervalo_promedio * 100 if intervalo_promedio > 0 else 0
    else:
        intervalo_promedio = 0
        desviacion_intervalos = 0
        coeficiente_variacion = 0
    
    return {
        'f_fundamental': f_fundamental,
        'relaciones_armonicas': relaciones_armonicas,
        'desviacion_promedio': np.mean(desviaciones),
        'intervalos': intervalos,
        'intervalo_promedio': intervalo_promedio,
        'desviacion_intervalos': desviacion_intervalos,
        'coeficiente_variacion': coeficiente_variacion,
        'es_simetrico': np.mean(desviaciones) < 5.0  # Criterio: < 5% de desviaci√≥n
    }

# ============================================================
# EJECUTAR AN√ÅLISIS DE SIMETR√çA
# ============================================================

print("üéº Analizando simetr√≠a entre modos de resonancia...")
print("=" * 50)

# Verificar dependencias
if 'resonancias_analizadas' not in locals():
    print("‚ùå ERROR: Variable 'resonancias_analizadas' no encontrada")
    print("\nüìã SOLUCI√ìN: Ejecuta las secciones en orden:")
    print("   1. Secci√≥n 1: Importaci√≥n de librer√≠as")
    print("   2. Secci√≥n 2: Carga de datos")
    print("   3. Secci√≥n 3: Detecci√≥n de picos")
    print("   4. Secci√≥n 4: An√°lisis de factor Q")
    print("   5. Secci√≥n 5: An√°lisis de simetr√≠a (esta secci√≥n)")
    print("\nüîÑ Ejecuta TODAS las secciones anteriores secuencialmente")
else:
    # Realizar an√°lisis
    simetria_data = analizar_simetria_modos(resonancias_analizadas)
    
    if simetria_data:
        print(f"üìä RESULTADOS DE SIMETR√çA:")
        print(f"   ‚Ä¢ Frecuencia fundamental: {simetria_data['f_fundamental']:.4f} GHz")
        print(f"   ‚Ä¢ N√∫mero de modos: {len(simetria_data['relaciones_armonicas'])}")
        
        # Tabla resumen concisa
        print(f"\nüéØ RELACIONES ARM√ìNICAS:")
        print(f"{'Modo':<6}{'Freq (GHz)':<12}{'Relaci√≥n':<10}{'Desv. (%)':<10}")
        print("-" * 38)
        
        for rel in simetria_data['relaciones_armonicas']:
            print(f"{rel['modo']:<6}{rel['frecuencia_ghz']:<12.4f}{rel['relacion_real']:<10.2f}{rel['desviacion_porcentual']:<10.1f}")
            
    else:
        print("‚ùå No se pudo realizar el an√°lisis (datos insuficientes)")

print("\n‚úÖ An√°lisis de simetr√≠a completado")
````

## üî¨ SECCI√ìN 6: Modelo Lorentziando Complejo

Implementamos **modelos Lorentzianos complejos** para extraer par√°metros precisos de cada resonancia usando **lmfit**.

````python
# ============================================================
# üî¨ AJUSTE INDIVIDUAL CON LMFIT - MODELO LORENTZIANO COMPLEJO
# ============================================================

def modelo_lorentziano_complejo(frecuencia, f0, Qc, amplitud, fase_deg, offset_real=0, offset_imag=0):
    """
    Modelo Lorentziano complejo para resonancia individual
    
    Par√°metros:
    - f0: frecuencia de resonancia (GHz)
    - Qc: factor de calidad cargado
    - amplitud: amplitud de la resonancia
    - fase_deg: fase en grados
    - offset_real, offset_imag: offsets para ajustar l√≠nea base
    """
    
    # Convertir fase a radianes
    fase_rad = np.radians(fase_deg)
    
    # Modelo Lorentziano complejo
    delta_f = frecuencia - f0
    denominador = 1 + 1j * 2 * Qc * delta_f / f0
    
    S21_model = amplitud * np.exp(1j * fase_rad) / denominador + (offset_real + 1j * offset_imag)
    
    return S21_model

def ajustar_resonancia_individual(frecuencias, s21_complejo, f0_inicial, ventana_ghz=0.2):
    """
    Ajusta una resonancia individual usando lmfit
    """
    
    # Seleccionar ventana alrededor de la resonancia
    mask = (frecuencias >= f0_inicial - ventana_ghz/2) & (frecuencias <= f0_inicial + ventana_ghz/2)
    freq_ventana = frecuencias[mask]
    s21_ventana = s21_complejo[mask]
    
    if len(freq_ventana) < 10:
        return None
    
    # Crear modelo lmfit
    modelo = Model(modelo_lorentziano_complejo, independent_vars=['frecuencia'])
    
    # Par√°metros iniciales
    params = modelo.make_params()
    
    # Estimaciones iniciales inteligentes
    s21_mag = np.abs(s21_ventana)
    s21_phase = np.angle(s21_ventana)
    
    idx_max = np.argmax(s21_mag)
    amplitud_inicial = np.max(s21_mag)
    fase_inicial = np.degrees(s21_phase[idx_max])
    
    # Estimar Q inicial
    nivel_3db = amplitud_inicial / np.sqrt(2)
    indices_3db = np.where(s21_mag >= nivel_3db)[0]
    if len(indices_3db) > 1:
        ancho_banda_inicial = freq_ventana[indices_3db[-1]] - freq_ventana[indices_3db[0]]
        Q_inicial = f0_inicial / ancho_banda_inicial if ancho_banda_inicial > 0 else 100
    else:
        Q_inicial = 100
    
    # Configurar par√°metros
    params['f0'].set(value=f0_inicial, min=f0_inicial-0.1, max=f0_inicial+0.1)
    params['Qc'].set(value=Q_inicial, min=10, max=1000)
    params['amplitud'].set(value=amplitud_inicial, min=0.1, max=2.0)
    params['fase_deg'].set(value=fase_inicial, min=-180, max=180)
    params['offset_real'].set(value=0, min=-0.5, max=0.5)
    params['offset_imag'].set(value=0, min=-0.5, max=0.5)
    
    # Preparar datos para ajuste (separar parte real e imaginaria)
    datos_ajuste = np.concatenate([s21_ventana.real, s21_ventana.imag])
    
    def residuales(params, frecuencia, datos):
        modelo_complejo = modelo_lorentziano_complejo(frecuencia, **params.valuesdict())
        modelo_separado = np.concatenate([modelo_complejo.real, modelo_complejo.imag])
        return datos - modelo_separado
    
    # Realizar ajuste
    try:
        resultado = lmfit.minimize(residuales, params, args=(freq_ventana, datos_ajuste), 
                                   method='leastsq')
        
        # Calcular modelo ajustado
        params_optimos = resultado.params
        s21_ajustado = modelo_lorentziano_complejo(freq_ventana, **params_optimos.valuesdict())
        
        # Calcular m√©tricas de calidad
        s21_modelo_separado = np.concatenate([s21_ajustado.real, s21_ajustado.imag])
        ss_res = np.sum((datos_ajuste - s21_modelo_separado)**2)
        ss_tot = np.sum((datos_ajuste - np.mean(datos_ajuste))**2)
        r_squared = 1 - (ss_res / ss_tot) if ss_tot > 0 else 0
        
        return {
            'exito': True,
            'resultado_lmfit': resultado,
            'parametros': params_optimos,
            'frecuencias_ajuste': freq_ventana,
            's21_medido': s21_ventana,
            's21_ajustado': s21_ajustado,
            'r_squared': r_squared,
            'chi_squared': resultado.chisqr,
            'chi_squared_reducido': resultado.redchi,
            # Par√°metros extra√≠dos
            'f0_ghz': params_optimos['f0'].value,
            'f0_error': params_optimos['f0'].stderr if params_optimos['f0'].stderr else 0,
            'Q_cargado': params_optimos['Qc'].value,
            'Q_error': params_optimos['Qc'].stderr if params_optimos['Qc'].stderr else 0,
            'amplitud': params_optimos['amplitud'].value,
            'fase_deg': params_optimos['fase_deg'].value,
            'fase_rad': np.radians(params_optimos['fase_deg'].value)
        }
        
    except Exception as e:
        print(f"Error en ajuste: {e}")
        return None

# ============================================================
# AJUSTAR CADA RESONANCIA INDIVIDUALMENTE
# ============================================================

ajustes_individuales = []

# Verificar que tenemos las variables necesarias
if 'resonancias_analizadas' in locals() and 'frecuencias' in locals() and 's21_complejo' in locals():
    
    for i, resonancia in enumerate(resonancias_analizadas):
        if resonancia['valido']:
            print(f"\nüéØ Ajustando Resonancia {resonancia['numero']} ({resonancia['frecuencia_ghz']:.4f} GHz)...")
            
            # Realizar ajuste
            ajuste = ajustar_resonancia_individual(
                frecuencias, 
                s21_complejo, 
                resonancia['frecuencia_ghz'],
                ventana_ghz=0.15  # Ventana de 150 MHz
            )
            
            if ajuste and ajuste['exito']:
                ajustes_individuales.append(ajuste)
                
                # Mostrar resultados
                print(f"      ‚Ä¢ f‚ÇÄ = {ajuste['f0_ghz']:.6f} ¬± {ajuste['f0_error']:.6f} GHz")
                print(f"      ‚Ä¢ Q = {ajuste['Q_cargado']:.1f} ¬± {ajuste['Q_error']:.1f}")
                print(f"      ‚Ä¢ R¬≤ = {ajuste['r_squared']:.6f}")
                print(f"      ‚Ä¢ œá¬≤·µ£ = {ajuste['chi_squared_reducido']:.4f}")
                
            else:
                print(f"   ‚ùå Ajuste fallido")
    
    print(f"\nüìä RESUMEN DE AJUSTES INDIVIDUALES:")
    print(f"   ‚Ä¢ Total resonancias: {len(resonancias_analizadas)}")
    print(f"   ‚Ä¢ Ajustes exitosos: {len(ajustes_individuales)}")
    
    if len(ajustes_individuales) > 0:
        r_squared_promedio = np.mean([ajuste['r_squared'] for ajuste in ajustes_individuales])
        print(f"   ‚Ä¢ R¬≤ promedio: {r_squared_promedio:.6f}")
        print(f"   ‚Ä¢ Calidad general: {'EXCELENTE' if r_squared_promedio > 0.99 else 'BUENA' if r_squared_promedio > 0.95 else 'REGULAR'}")
    
else:
    print("‚ùå Variables necesarias no encontradas. Ejecuta las secciones anteriores.")

print("\n‚úÖ Ajuste individual completado")
````

## üî¨ SECCI√ìN 6: MODELO COMPLEJO AVANZADO
### An√°lisis de Resonador con Par√°metros Complejos

Implementamos el modelo te√≥rico completo que considera:
- **Q interno (Q‚ÇÄ)**: p√©rdidas intr√≠nsecas del resonador
- **Q externo complejo (Q‚Çë)**: acoplamiento con puertos externos
- **Efectos de fase**: informaci√≥n completa del par√°metro S21
- **Modelo f√≠sico fundamentado**: basado en teor√≠a de resonadores acoplados

**Ventajas sobre el an√°lisis simple:**
- Mayor precisi√≥n en la extracci√≥n de par√°metros
- Separaci√≥n de p√©rdidas internas vs. acoplamientos
- Informaci√≥n de fase para caracterizaci√≥n completa
- Modelo v√°lido para resonadores d√©bil y fuertemente acoplados

````python
# üîß DEFINICI√ìN DEL MODELO COMPLEJO DE RESONADOR

def linear_resonator(f, f_0, Q, Q_e_real, Q_e_imag):
    """
    Modelo te√≥rico completo de resonador de transmisi√≥n (anillo)
    
    Par√°metros:
    - f: frecuencia [Hz]
    - f_0: frecuencia de resonancia [Hz]
    - Q: factor Q total (interno + externo) 
    - Q_e_real: parte real del Q externo (acoplamiento resistivo)
    - Q_e_imag: parte imaginaria del Q externo (acoplamiento reactivo)
    
    Retorna: S21 complejo
    """
    Q_e = Q_e_real + 1j * Q_e_imag
    # Modelo para resonador de transmisi√≥n (picos hacia arriba)
    denominador = 1 + 2j * Q * (f - f_0) / f_0
    return (Q * Q_e**-1) / denominador


class ResonatorModelComplejo(lmfit.model.Model):
    """Modelo lmfit para resonador con par√°metros complejos"""
    
    def __init__(self, *args, **kwargs):
        super().__init__(linear_resonator, *args, **kwargs)
        self.set_param_hint('Q', min=0)  # Q debe ser positivo
        self.set_param_hint('f_0', min=0)  # frecuencia debe ser positiva

    def guess(self, data, f=None, **kwargs):
        """Estimaci√≥n inicial inteligente de par√°metros"""
        verbose = kwargs.pop('verbose', None)
        if f is None:
            return
            
        # Encontrar el m√°ximo en magnitud (pico de transmisi√≥n para band-pass)
        argmax_s21 = np.abs(data).argmax()
        
        fmin = f.min()
        fmax = f.max()
        f_0_guess = f[argmax_s21]  # frecuencia de resonancia
        
        # Estimaci√≥n de Q basada en ancho de banda
        Q_min = 1.0 * (f_0_guess/(fmax-fmin))
        delta_f = np.diff(f)
        min_delta_f = delta_f[delta_f > 0].min()
        Q_max = f_0_guess/min_delta_f
        Q_guess = np.sqrt(Q_min * Q_max)
        
        # Estimaci√≥n de Q externo basada en altura de resonancia
        max_transmission = np.abs(data[argmax_s21])
        Q_e_real_guess = Q_guess * max_transmission if max_transmission > 0 else Q_guess
        
        if verbose:
            print(f"f_0_guess = {f_0_guess:.2e} Hz")
            print(f"Q_guess = {Q_guess:.1f}")
            print(f"Q_e_real_guess = {Q_e_real_guess:.1f}")
            print(f"Max transmission = {max_transmission:.3f}")
        
        params = self.make_params(
            Q=Q_guess, 
            Q_e_real=Q_e_real_guess, 
            Q_e_imag=0, 
            f_0=f_0_guess
        )
        
        # L√≠mites f√≠sicos
        params[f'{self.prefix}Q'].set(min=Q_min, max=Q_max)
        params[f'{self.prefix}f_0'].set(min=fmin, max=fmax)
        
        return lmfit.models.update_param_vals(params, self.prefix, **kwargs)

print("‚úÖ Modelo complejo de resonador definido correctamente")
````

````python
# üéØ AN√ÅLISIS COMPLEJO DE M√öLTIPLES RESONANCIAS

def analizar_resonancias_complejas(frecuencias, s21_complejo, picos_detectados, 
                                 ancho_ventana=0.1, verbose=True):
    """
    Analiza cada resonancia detectada usando el modelo complejo
    
    Par√°metros:
    - frecuencias: array de frecuencias [Hz]
    - s21_complejo: par√°metro S21 complejo
    - picos_detectados: lista de picos detectados con informaci√≥n
    - ancho_ventana: fracci√≥n del FSR para definir ventana de an√°lisis
    - verbose: mostrar informaci√≥n detallada
    
    Retorna: lista de resultados de ajuste complejo
    """
    
    resultados_complejos = []
    modelo = ResonatorModelComplejo()
    
    print(f"üî¨ Iniciando an√°lisis complejo de {len(picos_detectados)} resonancias...")
    print("=" * 70)
    
    for i, pico in enumerate(picos_detectados):
        f_resonancia = pico['frecuencia_ghz']  # Corregir nombre de clave
        indice_pico = pico['indice']
        
        # Definir ventana de an√°lisis alrededor del pico
        if i == 0:
            # Para la primera resonancia, evitar datos problem√°ticos cerca de 0 GHz
            f_inicio = max(frecuencias[0], f_resonancia - 0.2)  # M√°ximo 200 MHz hacia atr√°s
        else:
            f_prev = picos_detectados[i-1]['frecuencia_ghz']  # Corregir nombre de clave
            f_inicio = f_resonancia - (f_resonancia - f_prev) * ancho_ventana
            
        if i == len(picos_detectados) - 1:
            f_fin = frecuencias[-1]
        else:
            f_next = picos_detectados[i+1]['frecuencia_ghz']  # Corregir nombre de clave
            f_fin = f_resonancia + (f_next - f_resonancia) * ancho_ventana
        
        # Seleccionar datos en la ventana
        mask = (frecuencias >= f_inicio) & (frecuencias <= f_fin)
        f_local = frecuencias[mask]
        s21_local = s21_complejo[mask]
        
        if len(f_local) < 10:
            print(f"‚ö†Ô∏è  Resonancia {i+1}: Muy pocos puntos en ventana")
            continue
            
        try:
            # Convertir frecuencias de GHz a Hz para el modelo
            f_local_hz = f_local * 1e9
            
            # Estimaci√≥n inicial
            params_inicial = modelo.guess(s21_local, f=f_local_hz, verbose=verbose and i==0)
            
            # Ajuste complejo
            resultado = modelo.fit(s21_local, params_inicial, f=f_local_hz)
            
            # Extraer par√°metros f√≠sicos
            f0_fit = resultado.params['f_0'].value
            Q_fit = resultado.params['Q'].value
            Q_e_real_fit = resultado.params['Q_e_real'].value
            Q_e_imag_fit = resultado.params['Q_e_imag'].value
            
            # Calcular Q interno y externo
            Q_e_magnitude = np.sqrt(Q_e_real_fit**2 + Q_e_imag_fit**2)
            Q_internal = 1 / (1/Q_fit - 1/Q_e_magnitude)
            
            # Factor de acoplamiento
            kappa = Q_internal / Q_e_magnitude
            
            # Calcular R¬≤
            s21_fit = modelo.eval(resultado.params, f=f_local_hz)
            ss_res = np.sum(np.abs(s21_local - s21_fit)**2)
            ss_tot = np.sum(np.abs(s21_local - np.mean(s21_local))**2)
            r_squared = 1 - (ss_res / ss_tot)
            
            resultado_complejo = {
                'resonancia': i + 1,
                'frecuencia_original': f_resonancia,
                'f0_complejo': f0_fit,
                'Q_total': Q_fit,
                'Q_interno': Q_internal,
                'Q_externo_real': Q_e_real_fit,
                'Q_externo_imag': Q_e_imag_fit,
                'Q_externo_magnitud': Q_e_magnitude,
                'factor_acoplamiento': kappa,
                'r_squared_complejo': r_squared,
                'resultado_lmfit': resultado,
                'f_datos': f_local,
                's21_datos': s21_local,
                's21_ajuste': s21_fit,
                'ventana': (f_inicio, f_fin)
            }
            
            resultados_complejos.append(resultado_complejo)
            
            if verbose:
                print(f"‚úÖ Resonancia {i+1}:")
                print(f"   f‚ÇÄ = {f0_fit/1e9:.4f} GHz")
                print(f"   Q_total = {Q_fit:.1f}")
                print(f"   Q_interno = {Q_internal:.1f}")
                print(f"   Q_externo = {Q_e_magnitude:.1f}")
                print(f"   Œ∫ (acoplamiento) = {kappa:.3f}")
                print(f"   R¬≤ = {r_squared:.4f}")
                print(f"   Error cuadr√°tico = {resultado.chisqr:.2e}")
                print()
                
        except Exception as e:
            print(f"‚ùå Error en resonancia {i+1}: {str(e)}")
            continue
    
    print(f"üéâ An√°lisis complejo completado: {len(resultados_complejos)}/{len(picos_detectados)} resonancias analizadas")
    return resultados_complejos

print("‚úÖ Funci√≥n de an√°lisis complejo multi-resonancia definida")
````

## üåê SECCI√ìN 7: MODELO GLOBAL MULTI-MODO

### üéØ Objetivo
Implementar un **modelo global** que ajuste **todas las resonancias simult√°neamente**, considerando:
- **Efectos de interferencia** entre modos
- **Par√°metros f√≠sicos coherentes** entre resonancias  
- **Modelo te√≥ricamente fundamentado** para resonador de anillo
- **Separaci√≥n de Q interno vs. Q externo** por modo

### üî¨ Teor√≠a del Modelo Global
El modelo global multi-modo para un resonador de anillo considera la **suma coherente** de m√∫ltiples Lorentzianos:

$$S_{21}(\omega) = \sum_{n=1}^{N} \frac{A_n e^{i\phi_n}}{1 + 2iQ_n\frac{\omega - \omega_n}{\omega_n}} + B_{background}$$

Donde:
- $N$ = n√∫mero de modos resonantes
- $A_n, \phi_n$ = amplitud y fase del modo $n$
- $Q_n$ = factor de calidad del modo $n$
- $\omega_n$ = frecuencia de resonancia del modo $n$
- $B_{background}$ = t√©rmino de fondo (background) complejo

````python
# ============================================================
# üîß INICIALIZACI√ìN SEGURA DE VARIABLES GLOBALES
# ============================================================

# Esta celda inicializa variables para evitar errores de Pylance
# y garantizar que las funciones puedan acceder a las variables globales

# Verificar variables clave del an√°lisis anterior
variables_requeridas = {
    'frecuencias': None,
    's21_complejo': None, 
    'picos_detectados': None,
    'resultado_modelo_global': None,
    'parametros_modos_globales': None,
    'modelo_global_instancia': None
}

print("üîß Verificando variables globales necesarias...")

variables_faltantes = []
for var_name, default_value in variables_requeridas.items():
    if var_name not in globals():
        globals()[var_name] = default_value
        variables_faltantes.append(var_name)

if variables_faltantes:
    print(f"‚ö†Ô∏è  Variables inicializadas como None: {variables_faltantes}")
    print("üìã Para usar el modelo global, aseg√∫rate de ejecutar en orden:")
    print("   1. Secci√≥n 1: Importaci√≥n de librer√≠as")
    print("   2. Secci√≥n 2: Carga de datos")
    print("   3. Secci√≥n 3: Detecci√≥n de picos")
    print("   4. Secci√≥n 7: Modelo global (celdas siguientes)")
else:
    print("‚úÖ Todas las variables globales est√°n disponibles")

# Funci√≥n auxiliar para verificar estado del modelo global
def verificar_modelo_global():
    """Verifica si el modelo global ha sido ejecutado correctamente"""
    if globals().get('resultado_modelo_global') is None:
        return False, "Variable resultado_modelo_global no definida"
    
    resultado = globals().get('resultado_modelo_global')
    if not isinstance(resultado, dict):
        return False, "resultado_modelo_global no es un diccionario"
    
    if not resultado.get('exito', False):
        return False, "El modelo global no se ajust√≥ correctamente"
    
    return True, "Modelo global disponible y funcionando"

print("\n‚úÖ Inicializaci√≥n de variables globales completada")
````

````python
# ============================================================
# üåê IMPLEMENTACI√ìN DEL MODELO GLOBAL MULTI-MODO
# ============================================================

import numpy as np
import lmfit
from lmfit import Model, Parameters

class ModeloGlobalMultiModo:
    """
    Modelo global que ajusta m√∫ltiples resonancias simult√°neamente
    considerando efectos de interferencia y coherencia entre modos
    """
    
    def __init__(self, n_modos=4):
        """
        Inicializa el modelo global
        
        Par√°metros:
        - n_modos: n√∫mero de modos resonantes a ajustar
        """
        self.n_modos = n_modos
        self.modelo_lmfit = None
        self.parametros = None
        
    def funcion_modelo_global(self, frecuencia, **params):
        """
        Funci√≥n del modelo global multi-modo
        
        S21(f) = Œ£ [A_n * exp(i*œÜ_n) / (1 + 2i*Q_n*(f-f_n)/f_n)] + background
        """
        
        # Inicializar S21 con background complejo
        s21_global = params.get('bg_real', 0) + 1j * params.get('bg_imag', 0)
        
        # Sumar contribuci√≥n de cada modo
        for n in range(self.n_modos):
            # Par√°metros del modo n
            f_n = params[f'f{n}']          # Frecuencia de resonancia
            Q_n = params[f'Q{n}']          # Factor Q
            A_n = params[f'A{n}']          # Amplitud
            phi_n = params[f'phi{n}']      # Fase en radianes
            
            # T√©rmino Lorentziano para el modo n
            delta_f = frecuencia - f_n
            denominador = 1 + 2j * Q_n * delta_f / f_n
            
            # Contribuci√≥n compleja del modo n
            contribucion_n = A_n * np.exp(1j * phi_n) / denominador
            s21_global += contribucion_n
            
        return s21_global
    
    def crear_parametros_iniciales(self, frecuencias_picos, amplitudes_iniciales=None):
        """
        Crea par√°metros iniciales inteligentes basados en picos detectados
        """
        params = Parameters()
        
        # Par√°metros de background
        params.add('bg_real', value=0.0, min=-0.5, max=0.5)
        params.add('bg_imag', value=0.0, min=-0.5, max=0.5)
        
        # Par√°metros para cada modo
        for n in range(self.n_modos):
            if n < len(frecuencias_picos):
                f_inicial = frecuencias_picos[n]
                A_inicial = amplitudes_iniciales[n] if amplitudes_iniciales else 0.5
            else:
                # Si no hay suficientes picos, usar valores por defecto
                f_inicial = 1.0 + n * 0.2  # GHz espaciados
                A_inicial = 0.1
            
            # Frecuencia de resonancia
            params.add(f'f{n}', value=f_inicial, 
                      min=f_inicial*0.9, max=f_inicial*1.1)
            
            # Factor Q (t√≠pico para resonadores de anillo)
            params.add(f'Q{n}', value=100, min=10, max=1000)
            
            # Amplitud
            params.add(f'A{n}', value=A_inicial, min=0.01, max=2.0)
            
            # Fase (radianes)
            params.add(f'phi{n}', value=0.0, min=-np.pi, max=np.pi)
        
        return params
    
    def ajustar_modelo_global(self, frecuencias, s21_complejo, 
                             frecuencias_picos, verbose=True):
        """
        Ajusta el modelo global a los datos experimentales
        """
        
        # Estimar amplitudes iniciales basadas en los picos
        amplitudes_iniciales = []
        for f_pico in frecuencias_picos:
            # Encontrar √≠ndice m√°s cercano
            idx = np.argmin(np.abs(frecuencias - f_pico))
            amplitud = np.abs(s21_complejo[idx])
            amplitudes_iniciales.append(amplitud)
        
        # Crear par√°metros iniciales
        params_iniciales = self.crear_parametros_iniciales(
            frecuencias_picos, amplitudes_iniciales)
        
        # Crear modelo lmfit
        modelo = Model(self.funcion_modelo_global, independent_vars=['frecuencia'])
        
        # Preparar datos para ajuste (separar real e imaginario)
        datos_reales = s21_complejo.real
        datos_imaginarios = s21_complejo.imag
        datos_combinados = np.concatenate([datos_reales, datos_imaginarios])
        
        # Funci√≥n de residuales
        def residuales(params, frecuencia, datos):
            s21_modelo = self.funcion_modelo_global(frecuencia, **params.valuesdict())
            modelo_real = s21_modelo.real
            modelo_imag = s21_modelo.imag
            modelo_combinado = np.concatenate([modelo_real, modelo_imag])
            return datos - modelo_combinado
        
        if verbose:
            print("üî¨ Iniciando ajuste del modelo global multi-modo...")
            print(f"   ‚Ä¢ N√∫mero de modos: {self.n_modos}")
            print(f"   ‚Ä¢ Puntos de datos: {len(frecuencias)}")
            print(f"   ‚Ä¢ Par√°metros a ajustar: {len(params_iniciales)}")
            
        try:
            # Realizar ajuste
            resultado = lmfit.minimize(
                residuales, 
                params_iniciales, 
                args=(frecuencias, datos_combinados),
                method='leastsq'
            )
            
            # Calcular modelo ajustado
            s21_ajustado = self.funcion_modelo_global(
                frecuencias, **resultado.params.valuesdict())
            
            # Calcular m√©tricas de calidad
            modelo_combinado = np.concatenate([s21_ajustado.real, s21_ajustado.imag])
            ss_res = np.sum((datos_combinados - modelo_combinado)**2)
            ss_tot = np.sum((datos_combinados - np.mean(datos_combinados))**2)
            r_squared = 1 - (ss_res / ss_tot) if ss_tot > 0 else 0
            
            # Almacenar resultados
            self.modelo_lmfit = modelo
            self.parametros = resultado.params
            
            resultado_global = {
                'exito': True,
                'resultado_lmfit': resultado,
                'parametros': resultado.params,
                's21_ajustado': s21_ajustado,
                'r_squared': r_squared,
                'chi_squared': resultado.chisqr,
                'chi_squared_reducido': resultado.redchi,
                'n_parametros': len(params_iniciales),
                'n_datos': len(frecuencias)
            }
            
            if verbose:
                print(f"‚úÖ Ajuste completado:")
                print(f"   ‚Ä¢ R¬≤ = {r_squared:.6f}")
                print(f"   ‚Ä¢ œá¬≤ reducido = {resultado.redchi:.4f}")
                print(f"   ‚Ä¢ √âxito de convergencia: {resultado.success}")
                
            return resultado_global
            
        except Exception as e:
            if verbose:
                print(f"‚ùå Error en ajuste global: {e}")
            return {'exito': False, 'error': str(e)}
    
    def extraer_parametros_fisicos(self, resultado_global):
        """
        Extrae par√°metros f√≠sicos de cada modo del ajuste global
        """
        if not resultado_global['exito']:
            return None
            
        parametros_fisicos = []
        params = resultado_global['parametros']
        
        for n in range(self.n_modos):
            try:
                modo_params = {
                    'modo': n + 1,
                    'frecuencia_ghz': params[f'f{n}'].value,
                    'frecuencia_error': params[f'f{n}'].stderr if params[f'f{n}'].stderr else 0,
                    'Q_total': params[f'Q{n}'].value,
                    'Q_error': params[f'Q{n}'].stderr if params[f'Q{n}'].stderr else 0,
                    'amplitud': params[f'A{n}'].value,
                    'amplitud_error': params[f'A{n}'].stderr if params[f'A{n}'].stderr else 0,
                    'fase_rad': params[f'phi{n}'].value,
                    'fase_deg': np.degrees(params[f'phi{n}'].value),
                    'fase_error': params[f'phi{n}'].stderr if params[f'phi{n}'].stderr else 0
                }
                parametros_fisicos.append(modo_params)
            except KeyError:
                continue
                
        return parametros_fisicos

print("‚úÖ Clase ModeloGlobalMultiModo implementada correctamente")
````

````python
# ============================================================
# üöÄ APLICACI√ìN DEL MODELO GLOBAL CON DATOS REALES
# ============================================================

# Verificar que tenemos las variables necesarias del an√°lisis anterior
variables_requeridas = ['frecuencias', 's21_complejo', 'picos_detectados']
variables_faltantes = [var for var in variables_requeridas if var not in globals()]

if variables_faltantes:
    print("‚ùå ERROR: Variables faltantes del an√°lisis anterior:")
    for var in variables_faltantes:
        print(f"   ‚Ä¢ {var}")
    print("\nüìã SOLUCI√ìN: Ejecuta las secciones anteriores en orden:")
    print("   1. Secci√≥n 1: Importaci√≥n de librer√≠as")
    print("   2. Secci√≥n 2: Carga de datos") 
    print("   3. Secci√≥n 3: Detecci√≥n de picos")
    print("   4. Luego ejecuta esta secci√≥n")
else:
    print("‚úÖ Variables necesarias encontradas. Procediendo con modelo global...")
    
    # Extraer frecuencias de picos detectados
    if len(picos_detectados) > 0:
        frecuencias_picos = [pico['frecuencia_ghz'] for pico in picos_detectados]
        n_picos = len(frecuencias_picos)
        
        print(f"\nüéØ CONFIGURACI√ìN DEL MODELO GLOBAL:")
        print(f"   ‚Ä¢ N√∫mero de resonancias detectadas: {n_picos}")
        print(f"   ‚Ä¢ Frecuencias de picos: {[f'{f:.4f} GHz' for f in frecuencias_picos]}")
        print(f"   ‚Ä¢ Rango de datos: {frecuencias[0]:.2f} - {frecuencias[-1]:.2f} GHz")
        print(f"   ‚Ä¢ Puntos de datos: {len(frecuencias):,}")
        
        # Crear instancia del modelo global
        modelo_global = ModeloGlobalMultiModo(n_modos=n_picos)
        
        print(f"\nüî¨ Ejecutando ajuste del modelo global...")
        print("=" * 60)
        
        # Ajustar modelo global
        resultado_global = modelo_global.ajustar_modelo_global(
            frecuencias=frecuencias,
            s21_complejo=s21_complejo,
            frecuencias_picos=frecuencias_picos,
            verbose=True
        )
        
        if resultado_global['exito']:
            print(f"\nüéâ ¬°AJUSTE GLOBAL EXITOSO!")
            print("=" * 60)
            
            # Extraer par√°metros f√≠sicos de cada modo
            parametros_modos = modelo_global.extraer_parametros_fisicos(resultado_global)
            
            if parametros_modos:
                print(f"\nüìä PAR√ÅMETROS EXTRA√çDOS POR MODO:")
                print("-" * 80)
                print(f"{'Modo':<6}{'f‚ÇÄ (GHz)':<12}{'Q total':<10}{'Amplitud':<12}{'Fase (¬∞)':<10}")
                print("-" * 80)
                
                for params in parametros_modos:
                    modo = params['modo']
                    f0 = params['frecuencia_ghz']
                    Q = params['Q_total']
                    A = params['amplitud']
                    fase = params['fase_deg']
                    
                    print(f"{modo:<6}{f0:<12.6f}{Q:<10.1f}{A:<12.4f}{fase:<10.1f}")
                
                print("-" * 80)
                print(f"Calidad del ajuste global: R¬≤ = {resultado_global['r_squared']:.6f}")
                
                # Almacenar resultados para an√°lisis posterior
                globals()['resultado_modelo_global'] = resultado_global
                globals()['parametros_modos_globales'] = parametros_modos
                globals()['modelo_global_instancia'] = modelo_global
                
                print(f"\n‚úÖ Resultados almacenados en variables globales:")
                print(f"   ‚Ä¢ resultado_modelo_global")
                print(f"   ‚Ä¢ parametros_modos_globales") 
                print(f"   ‚Ä¢ modelo_global_instancia")
                
            else:
                print("‚ö†Ô∏è No se pudieron extraer par√°metros f√≠sicos")
                
        else:
            print(f"‚ùå Error en ajuste global: {resultado_global.get('error', 'Error desconocido')}")
            
    else:
        print("‚ùå No se encontraron picos detectados para el modelo global")
````

````python
# ============================================================
# üìä VISUALIZACI√ìN DEL MODELO GLOBAL vs. DATOS EXPERIMENTALES
# ============================================================

def graficar_modelo_global():
    """
    Crea gr√°ficos comparativos del modelo global vs. datos experimentales
    """
    
    # Verificar que tenemos los resultados del modelo global
    if 'resultado_modelo_global' not in globals():
        print("‚ùå Primero ejecuta la celda del modelo global")
        return
    
    # Obtener la variable global de forma segura
    resultado_global = globals().get('resultado_modelo_global')
    if resultado_global is None or not resultado_global.get('exito', False):
        print("‚ùå El modelo global no se ajust√≥ correctamente")
        return
    
    # Datos del modelo global
    s21_global = resultado_global['s21_ajustado']
    r_squared = resultado_global['r_squared']
    
    # Crear figura con m√∫ltiples subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    
    # =============================================================
    # GR√ÅFICO 1: MAGNITUD - COMPARACI√ìN GLOBAL
    # =============================================================
    
    # Datos experimentales
    mag_experimental = 20 * np.log10(np.abs(s21_complejo))
    mag_global = 20 * np.log10(np.abs(s21_global))
    
    ax1.plot(frecuencias, mag_experimental, 'b-', alpha=0.7, linewidth=1.5, 
             label='Datos experimentales')
    ax1.plot(frecuencias, mag_global, 'r-', linewidth=2, 
             label=f'Modelo Global (R¬≤={r_squared:.4f})')
    
    # Marcar resonancias detectadas
    if 'picos_detectados' in globals():
        for i, pico in enumerate(picos_detectados):
            ax1.axvline(pico['frecuencia_ghz'], color='gray', alpha=0.5, linestyle='--')
            ax1.text(pico['frecuencia_ghz'], max(mag_experimental), f'M{i+1}', 
                    ha='center', va='bottom', fontsize=8)
    
    ax1.set_xlabel('Frecuencia (GHz)')
    ax1.set_ylabel('|S21| (dB)')
    ax1.set_title('Magnitud S21: Modelo Global vs. Experimental')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # =============================================================
    # GR√ÅFICO 2: FASE - COMPARACI√ìN GLOBAL  
    # =============================================================
    
    fase_experimental = np.degrees(np.angle(s21_complejo))
    fase_global = np.degrees(np.angle(s21_global))
    
    ax2.plot(frecuencias, fase_experimental, 'b-', alpha=0.7, linewidth=1.5,
             label='Datos experimentales')
    ax2.plot(frecuencias, fase_global, 'r-', linewidth=2,
             label='Modelo Global')
    
    ax2.set_xlabel('Frecuencia (GHz)')
    ax2.set_ylabel('Fase S21 (grados)')
    ax2.set_title('Fase S21: Modelo Global vs. Experimental')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # =============================================================
    # GR√ÅFICO 3: PLANO COMPLEJO - TRAYECTORIA COMPLETA
    # =============================================================
    
    ax3.plot(s21_complejo.real, s21_complejo.imag, 'b-', alpha=0.6, linewidth=1,
             label='Datos experimentales')
    ax3.plot(s21_global.real, s21_global.imag, 'r-', linewidth=2,
             label='Modelo Global')
    
    # Marcar puntos de inicio y final
    ax3.scatter(s21_complejo.real[0], s21_complejo.imag[0], 
               color='green', s=50, marker='s', zorder=5, label='Inicio')
    ax3.scatter(s21_complejo.real[-1], s21_complejo.imag[-1], 
               color='red', s=50, marker='^', zorder=5, label='Final')
    
    ax3.set_xlabel('Parte Real S21')
    ax3.set_ylabel('Parte Imaginaria S21')
    ax3.set_title('Plano Complejo: Trayectoria Global')
    ax3.grid(True, alpha=0.3)
    ax3.axis('equal')
    ax3.legend()
    
    # =============================================================
    # GR√ÅFICO 4: RESIDUALES DEL AJUSTE
    # =============================================================
    
    # Calcular residuales
    residuales_mag = mag_experimental - mag_global
    residuales_fase = fase_experimental - fase_global
    
    ax4.plot(frecuencias, residuales_mag, 'g-', linewidth=1.5, alpha=0.8,
             label='Residuales Magnitud (dB)')
    ax4_twin = ax4.twinx()
    ax4_twin.plot(frecuencias, residuales_fase, 'm-', linewidth=1.5, alpha=0.8,
                  label='Residuales Fase (¬∞)')
    
    ax4.set_xlabel('Frecuencia (GHz)')
    ax4.set_ylabel('Residual Magnitud (dB)', color='g')
    ax4_twin.set_ylabel('Residual Fase (¬∞)', color='m')
    ax4.set_title('Residuales del Ajuste Global')
    ax4.grid(True, alpha=0.3)
    
    # Leyendas combinadas
    lines1, labels1 = ax4.get_legend_handles_labels()
    lines2, labels2 = ax4_twin.get_legend_handles_labels()
    ax4.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
    
    plt.tight_layout()
    plt.show()
    
    # =============================================================
    # ESTAD√çSTICAS DEL AJUSTE
    # =============================================================
    
    print(f"\nüìà ESTAD√çSTICAS DEL MODELO GLOBAL:")
    print("=" * 50)
    print(f"R¬≤ global:              {r_squared:.6f}")
    print(f"œá¬≤ reducido:            {resultado_global['chi_squared_reducido']:.4f}")
    print(f"RMS residual magnitud:  {np.sqrt(np.mean(residuales_mag**2)):.4f} dB")
    print(f"RMS residual fase:      {np.sqrt(np.mean(residuales_fase**2)):.2f}¬∞")
    print(f"N√∫mero de par√°metros:   {resultado_global['n_parametros']}")
    print(f"Puntos de datos:        {resultado_global['n_datos']:,}")
    
    # Calidad del ajuste
    if r_squared > 0.99:
        calidad = "EXCELENTE"
    elif r_squared > 0.95:
        calidad = "BUENA"
    elif r_squared > 0.90:
        calidad = "REGULAR"
    else:
        calidad = "POBRE"
        
    print(f"Calidad del ajuste:     {calidad}")
    
    return True

# Ejecutar visualizaci√≥n si tenemos los datos
resultado_global_temp = globals().get('resultado_modelo_global')
if resultado_global_temp is not None and resultado_global_temp.get('exito', False):
    print("üìä Generando gr√°ficos del modelo global...")
    graficar_modelo_global()
else:
    print("‚ö†Ô∏è Ejecuta primero la celda del modelo global para ver las visualizaciones")
````

````python
# ============================================================
# üìä VERSI√ìN MEJORADA - VISUALIZACI√ìN DEL MODELO GLOBAL
# ============================================================

def graficar_modelo_global_seguro():
    """
    Versi√≥n mejorada que maneja variables globales de forma segura
    """
    
    # Verificar estado del modelo global
    estado_ok, mensaje = verificar_modelo_global()
    if not estado_ok:
        print(f"‚ùå {mensaje}")
        print("üìã Ejecuta primero las celdas del modelo global en orden")
        return False
    
    # Obtener variables globales de forma segura
    try:
        resultado_global = globals()['resultado_modelo_global']
        s21_complejo = globals()['s21_complejo']
        frecuencias = globals()['frecuencias']
        picos_detectados = globals().get('picos_detectados', [])
        
        # Datos del modelo global
        s21_global = resultado_global['s21_ajustado']
        r_squared = resultado_global['r_squared']
        
    except KeyError as e:
        print(f"‚ùå Variable faltante: {e}")
        return False
    except Exception as e:
        print(f"‚ùå Error al acceder a datos: {e}")
        return False
    
    # Crear figura con m√∫ltiples subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    
    # =============================================================
    # GR√ÅFICO 1: MAGNITUD - COMPARACI√ìN GLOBAL
    # =============================================================
    
    mag_experimental = 20 * np.log10(np.abs(s21_complejo))
    mag_global = 20 * np.log10(np.abs(s21_global))
    
    ax1.plot(frecuencias, mag_experimental, 'b-', alpha=0.7, linewidth=1.5, 
             label='Datos experimentales')
    ax1.plot(frecuencias, mag_global, 'r-', linewidth=2, 
             label=f'Modelo Global (R¬≤={r_squared:.4f})')
    
    # Marcar resonancias detectadas
    for i, pico in enumerate(picos_detectados):
        ax1.axvline(pico['frecuencia_ghz'], color='gray', alpha=0.5, linestyle='--')
        ax1.text(pico['frecuencia_ghz'], max(mag_experimental), f'M{i+1}', 
                ha='center', va='bottom', fontsize=8)
    
    ax1.set_xlabel('Frecuencia (GHz)')
    ax1.set_ylabel('|S21| (dB)')
    ax1.set_title('Magnitud S21: Modelo Global vs. Experimental')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # =============================================================
    # GR√ÅFICO 2: FASE - COMPARACI√ìN GLOBAL  
    # =============================================================
    
    fase_experimental = np.degrees(np.angle(s21_complejo))
    fase_global = np.degrees(np.angle(s21_global))
    
    ax2.plot(frecuencias, fase_experimental, 'b-', alpha=0.7, linewidth=1.5,
             label='Datos experimentales')
    ax2.plot(frecuencias, fase_global, 'r-', linewidth=2,
             label='Modelo Global')
    
    ax2.set_xlabel('Frecuencia (GHz)')
    ax2.set_ylabel('Fase S21 (grados)')
    ax2.set_title('Fase S21: Modelo Global vs. Experimental')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # =============================================================
    # GR√ÅFICO 3: PLANO COMPLEJO - TRAYECTORIA COMPLETA
    # =============================================================
    
    ax3.plot(s21_complejo.real, s21_complejo.imag, 'b-', alpha=0.6, linewidth=1,
             label='Datos experimentales')
    ax3.plot(s21_global.real, s21_global.imag, 'r-', linewidth=2,
             label='Modelo Global')
    
    # Marcar puntos de inicio y final
    ax3.scatter(s21_complejo.real[0], s21_complejo.imag[0], 
               color='green', s=50, marker='s', zorder=5, label='Inicio')
    ax3.scatter(s21_complejo.real[-1], s21_complejo.imag[-1], 
               color='red', s=50, marker='^', zorder=5, label='Final')
    
    ax3.set_xlabel('Parte Real S21')
    ax3.set_ylabel('Parte Imaginaria S21')
    ax3.set_title('Plano Complejo: Trayectoria Global')
    ax3.grid(True, alpha=0.3)
    ax3.axis('equal')
    ax3.legend()
    
    # =============================================================
    # GR√ÅFICO 4: RESIDUALES DEL AJUSTE
    # =============================================================
    
    residuales_mag = mag_experimental - mag_global
    residuales_fase = fase_experimental - fase_global
    
    ax4.plot(frecuencias, residuales_mag, 'g-', linewidth=1.5, alpha=0.8,
             label='Residuales Magnitud (dB)')
    ax4_twin = ax4.twinx()
    ax4_twin.plot(frecuencias, residuales_fase, 'm-', linewidth=1.5, alpha=0.8,
                  label='Residuales Fase (¬∞)')
    
    ax4.set_xlabel('Frecuencia (GHz)')
    ax4.set_ylabel('Residual Magnitud (dB)', color='g')
    ax4_twin.set_ylabel('Residual Fase (¬∞)', color='m')
    ax4.set_title('Residuales del Ajuste Global')
    ax4.grid(True, alpha=0.3)
    
    # Leyendas combinadas
    lines1, labels1 = ax4.get_legend_handles_labels()
    lines2, labels2 = ax4_twin.get_legend_handles_labels()
    ax4.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
    
    plt.tight_layout()
    plt.show()
    
    # =============================================================
    # ESTAD√çSTICAS DEL AJUSTE
    # =============================================================
    
    print(f"\nüìà ESTAD√çSTICAS DEL MODELO GLOBAL:")
    print("=" * 50)
    print(f"R¬≤ global:              {r_squared:.6f}")
    print(f"œá¬≤ reducido:            {resultado_global['chi_squared_reducido']:.4f}")
    print(f"RMS residual magnitud:  {np.sqrt(np.mean(residuales_mag**2)):.4f} dB")
    print(f"RMS residual fase:      {np.sqrt(np.mean(residuales_fase**2)):.2f}¬∞")
    print(f"N√∫mero de par√°metros:   {resultado_global['n_parametros']}")
    print(f"Puntos de datos:        {resultado_global['n_datos']:,}")
    
    # Calidad del ajuste
    if r_squared > 0.99:
        calidad = "EXCELENTE"
    elif r_squared > 0.95:
        calidad = "BUENA"
    elif r_squared > 0.90:
        calidad = "REGULAR"
    else:
        calidad = "POBRE"
        
    print(f"Calidad del ajuste:     {calidad}")
    
    return True

# Ejecutar visualizaci√≥n de forma segura
print("üìä Intentando generar gr√°ficos del modelo global...")
resultado_viz = graficar_modelo_global_seguro()

if not resultado_viz:
    print("\nüí° INSTRUCCIONES PARA EJECUTAR EL MODELO GLOBAL:")
    print("   1. Ejecuta la celda de 'Importaci√≥n de librer√≠as'")
    print("   2. Ejecuta la celda de 'Carga de datos'") 
    print("   3. Ejecuta la celda de 'Detecci√≥n de picos'")
    print("   4. Ejecuta la celda de 'Implementaci√≥n del modelo global'")
    print("   5. Ejecuta la celda de 'Aplicaci√≥n con datos reales'")
    print("   6. Finalmente ejecuta esta celda de visualizaci√≥n")
````

````python
# ============================================================
# üîç VERSI√ìN MEJORADA - AN√ÅLISIS COMPARATIVO SEGURO
# ============================================================

def comparar_modelos_seguro():
    """
    Versi√≥n mejorada que maneja variables globales de forma segura
    """
    
    # Verificar estado del modelo global
    estado_ok, mensaje = verificar_modelo_global()
    if not estado_ok:
        print(f"‚ùå {mensaje}")
        return False
    
    # Obtener variables globales de forma segura
    try:
        resultado_global = globals()['resultado_modelo_global']
        parametros_globales = globals()['parametros_modos_globales']
        ajustes_individuales = globals().get('ajustes_individuales', [])
        
        if parametros_globales is None:
            print("‚ùå parametros_modos_globales no est√° definido")
            return False
            
    except KeyError as e:
        print(f"‚ùå Variable faltante: {e}")
        return False
    except Exception as e:
        print(f"‚ùå Error al acceder a datos: {e}")
        return False
    
    print("üîç COMPARACI√ìN MODELO INDIVIDUAL vs. MODELO GLOBAL")
    print("=" * 70)
    
    # Encabezado de la tabla
    print(f"{'Modo':<6}{'M√©todo':<12}{'f‚ÇÄ (GHz)':<12}{'Q total':<10}{'Amplitud':<12}{'R¬≤':<8}")
    print("-" * 70)
    
    # Obtener datos del modelo global
    r2_global = resultado_global['r_squared']
    
    for i, params_global in enumerate(parametros_globales):
        modo_num = params_global['modo']
        
        # Datos del modelo global
        f0_global = params_global['frecuencia_ghz']
        Q_global = params_global['Q_total']
        A_global = params_global['amplitud']
        
        print(f"{modo_num:<6}{'Global':<12}{f0_global:<12.6f}{Q_global:<10.1f}{A_global:<12.4f}{r2_global:<8.4f}")
        
        # Buscar datos del modelo individual correspondiente (si existe)
        if len(ajustes_individuales) > i:
            ajuste_ind = ajustes_individuales[i]
            f0_ind = ajuste_ind['f0_ghz']
            Q_ind = ajuste_ind['Q_cargado']
            A_ind = ajuste_ind['amplitud']
            r2_ind = ajuste_ind['r_squared']
            
            print(f"{modo_num:<6}{'Individual':<12}{f0_ind:<12.6f}{Q_ind:<10.1f}{A_ind:<12.4f}{r2_ind:<8.4f}")
            
            # Calcular diferencias relativas
            diff_f0 = abs(f0_global - f0_ind) / f0_ind * 100
            diff_Q = abs(Q_global - Q_ind) / Q_ind * 100
            
            print(f"{'':<6}{'Diff (%)':<12}{diff_f0:<12.2f}{diff_Q:<10.2f}{'':<12}{'':<8}")
        else:
            print(f"{modo_num:<6}{'Individual':<12}{'N/A':<12}{'N/A':<10}{'N/A':<12}{'N/A':<8}")
        
        print()  # L√≠nea en blanco entre modos
    
    print("-" * 70)
    
    # Resumen estad√≠stico
    if len(ajustes_individuales) > 0:
        r2_individual_promedio = np.mean([aj['r_squared'] for aj in ajustes_individuales])
        print(f"\nüìä RESUMEN ESTAD√çSTICO:")
        print(f"   ‚Ä¢ R¬≤ modelo individual (promedio): {r2_individual_promedio:.6f}")
        print(f"   ‚Ä¢ R¬≤ modelo global:                 {r2_global:.6f}")
        
        # Ventajas del modelo global
        if r2_global > r2_individual_promedio:
            print(f"   ‚Ä¢ ‚úÖ Modelo global tiene mejor ajuste general")
        else:
            print(f"   ‚Ä¢ ‚ö†Ô∏è Modelo individual tiene mejor ajuste promedio")
    else:
        print(f"\nüìä RESUMEN ESTAD√çSTICO:")
        print(f"   ‚Ä¢ R¬≤ modelo global: {r2_global:.6f}")
        print(f"   ‚Ä¢ Modelos individuales: No disponibles")
    
    print(f"\nüéØ VENTAJAS DEL MODELO GLOBAL:")
    print(f"   ‚úÖ Considera interferencia entre modos")
    print(f"   ‚úÖ Par√°metros f√≠sicamente coherentes")
    print(f"   ‚úÖ Ajuste simult√°neo de todas las resonancias")
    print(f"   ‚úÖ Modelo te√≥ricamente fundamentado")
    print(f"   ‚úÖ Reduce sobreajuste de par√°metros individuales")
    
    print(f"\nüéØ CARACTER√çSTICAS DEL MODELO INDIVIDUAL:")
    print(f"   ‚Ä¢ Optimizaci√≥n local por resonancia")
    print(f"   ‚Ä¢ Puede sobreajustar datos locales")
    print(f"   ‚Ä¢ No considera efectos de interferencia")
    print(f"   ‚Ä¢ √ötil para an√°lisis de resonancias aisladas")
    
    return True

# Ejecutar comparaci√≥n de forma segura
print("üîç Intentando comparar modelos...")
resultado_comp = comparar_modelos_seguro()

if not resultado_comp:
    print("\nüí° Para ejecutar la comparaci√≥n completa:")
    print("   1. Ejecuta todas las secciones anteriores en orden")
    print("   2. Especialmente el modelo global (Secci√≥n 7)")
    print("   3. Y tambi√©n los modelos individuales (Secci√≥n 6)")
    print("   4. Luego ejecuta esta celda de comparaci√≥n")
````

````python
# ============================================================
# üîç AN√ÅLISIS COMPARATIVO: MODELO INDIVIDUAL vs. MODELO GLOBAL
# ============================================================

def comparar_modelos():
    """
    Compara los resultados del modelo individual vs. modelo global
    """
    
    # Verificar que tenemos ambos resultados
    variables_necesarias = ['parametros_modos_globales', 'resultado_modelo_global']
    
    if not all(var in globals() for var in variables_necesarias):
        print("‚ùå Faltan resultados del modelo global. Ejecuta las celdas anteriores.")
        return
    
    # Obtener variables globales de forma segura
    resultado_global = globals().get('resultado_modelo_global')
    parametros_globales = globals().get('parametros_modos_globales')
    
    if resultado_global is None or parametros_globales is None:
        print("‚ùå Faltan datos del modelo global.")
        return
    
    print("üîç COMPARACI√ìN MODELO INDIVIDUAL vs. MODELO GLOBAL")
    print("=" * 70)
    
    # Encabezado de la tabla
    print(f"{'Modo':<6}{'M√©todo':<12}{'f‚ÇÄ (GHz)':<12}{'Q total':<10}{'Amplitud':<12}{'R¬≤':<8}")
    print("-" * 70)
    
    # Obtener datos del modelo global
    r2_global = resultado_global['r_squared']
    
    for i, params_global in enumerate(parametros_globales):
        modo_num = params_global['modo']
        
        # Datos del modelo global
        f0_global = params_global['frecuencia_ghz']
        Q_global = params_global['Q_total']
        A_global = params_global['amplitud']
        
        print(f"{modo_num:<6}{'Global':<12}{f0_global:<12.6f}{Q_global:<10.1f}{A_global:<12.4f}{r2_global:<8.4f}")
        
        # Buscar datos del modelo individual correspondiente (si existe)
        if 'ajustes_individuales' in globals() and i < len(ajustes_individuales):
            ajuste_ind = ajustes_individuales[i]
            f0_ind = ajuste_ind['f0_ghz']
            Q_ind = ajuste_ind['Q_cargado']
            A_ind = ajuste_ind['amplitud']
            r2_ind = ajuste_ind['r_squared']
            
            print(f"{modo_num:<6}{'Individual':<12}{f0_ind:<12.6f}{Q_ind:<10.1f}{A_ind:<12.4f}{r2_ind:<8.4f}")
            
            # Calcular diferencias relativas
            diff_f0 = abs(f0_global - f0_ind) / f0_ind * 100
            diff_Q = abs(Q_global - Q_ind) / Q_ind * 100
            
            print(f"{'':<6}{'Diff (%)':<12}{diff_f0:<12.2f}{diff_Q:<10.2f}{'':<12}{'':<8}")
        
        print()  # L√≠nea en blanco entre modos
    
    print("-" * 70)
    
    # Resumen estad√≠stico
    if 'ajustes_individuales' in globals():
        print(f"\nüìä RESUMEN ESTAD√çSTICO:")
        print(f"   ‚Ä¢ R¬≤ modelo individual (promedio): {np.mean([aj['r_squared'] for aj in ajustes_individuales]):.6f}")
        print(f"   ‚Ä¢ R¬≤ modelo global:                 {r2_global:.6f}")
        
        # Ventajas del modelo global
        if r2_global > np.mean([aj['r_squared'] for aj in ajustes_individuales]):
            print(f"   ‚Ä¢ ‚úÖ Modelo global tiene mejor ajuste general")
        else:
            print(f"   ‚Ä¢ ‚ö†Ô∏è Modelo individual tiene mejor ajuste promedio")
    
    print(f"\nüéØ VENTAJAS DEL MODELO GLOBAL:")
    print(f"   ‚úÖ Considera interferencia entre modos")
    print(f"   ‚úÖ Par√°metros f√≠sicamente coherentes")
    print(f"   ‚úÖ Ajuste simult√°neo de todas las resonancias")
    print(f"   ‚úÖ Modelo te√≥ricamente fundamentado")
    print(f"   ‚úÖ Reduce sobreajuste de par√°metros individuales")
    
    print(f"\nüéØ CARACTER√çSTICAS DEL MODELO INDIVIDUAL:")
    print(f"   ‚Ä¢ Optimizaci√≥n local por resonancia")
    print(f"   ‚Ä¢ Puede sobreajustar datos locales")
    print(f"   ‚Ä¢ No considera efectos de interferencia")
    print(f"   ‚Ä¢ √ötil para an√°lisis de resonancias aisladas")
    
    return True

# Ejecutar comparaci√≥n si tenemos los datos
if 'parametros_modos_globales' in globals():
    comparar_modelos()
else:
    print("‚ö†Ô∏è Ejecuta primero el modelo global para hacer la comparaci√≥n")
````
