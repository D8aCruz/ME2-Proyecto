```python
# ============================================================
# üì¶ IMPORTACI√ìN DE LIBRER√çAS ESENCIALES
# ============================================================
import os
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import signal              # usado para find_peaks
from scipy.optimize import minimize   # usado en algunos ajustes
import lmfit                          # n√∫cleo de los ajustes no lineales
from lmfit import Model               # para definir modelos de ajuste

print("‚úÖ Librer√≠as importadas correctamente")
print(f"üì¶ lmfit version: {lmfit.__version__}")
print(f"üìä numpy version: {np.__version__}")
print(f"üêº pandas version: {pd.__version__}")
```

```python
# ============================================================
# üé® CONFIGURACI√ìN DE ESTILO, CONSTANTES Y COLORES
# ============================================================
warnings.filterwarnings('ignore', category=RuntimeWarning)

plt.style.use('default')
plt.rcParams.update({
    'figure.figsize': (12, 8),
    'font.size': 11,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'axes.formatter.useoffset': False,
})
np.set_printoptions(precision=4, suppress=True)

COLORES = {
    'resonancia':   '#E74C3C',  # rojo para marcar f0
    'ajuste':       '#8FBC8F',  # verde para curvas de modelo/fit
    'datos':        '#3498DB',  # azul para datos
    'fondo':        '#95A5A6',  # gris
    'destacado':    '#F39C12',  # naranja para resaltar
    'comparacion':  '#9B59B6',  # morado para comparativas
    'incertidumbre':'#E67E22',  # naranja oscuro para barras de error
}

ARCHIVO_S2P = 'Rogers-1GHz.s2p'  # archivo Touchstone de trabajo

print("üé® Estilo aplicado y paleta definida")
print("üìÅ Archivo objetivo:", ARCHIVO_S2P)
```

```python
# ============================================================
# üìÇ FUNCI√ìN DE CARGA DE DATOS S21 - ROGERS 1GHz
# ============================================================

def cargar_datos_rogers_s2p():
    """
    Carga datos S21 del archivo Rogers-1GHz.s2p en formato Touchstone.
    """
    if not os.path.exists(ARCHIVO_S2P):
        print(f"‚ùå Archivo no encontrado: {ARCHIVO_S2P}")
        return None
    
    try:
        data = []
        with open(ARCHIVO_S2P, 'r') as file:
            for line in file:
                line = line.strip()
                if line.startswith(('!', '#')) or not line:
                    continue
                values = [float(x) for x in line.split()]
                if len(values) == 9:
                    data.append(values)

        data = np.array(data)
        f_GHz = data[:, 0]
        s21_mag = data[:, 5]
        s21_ang_deg = data[:, 6]

        S21 = s21_mag * np.exp(1j * np.deg2rad(s21_ang_deg))
        S21_db = 20 * np.log10(s21_mag + 1e-15)

        df = pd.DataFrame({
            "f_GHz": f_GHz,
            "S21_db": S21_db,
            "S21_mag": s21_mag,
            "S21_ang": s21_ang_deg,
            "S21": S21
        })

        return df
    
    except Exception as e:
        print(f"‚ö†Ô∏è Error al procesar {ARCHIVO_S2P}: {e}")
        return None
```

```python
# ============================================================
# üìÇ CARGA Y VISUALIZACI√ìN INICIAL DE DATOS S21 - ROGERS 1GHz
# ============================================================

df = cargar_datos_rogers_s2p()

if df is not None:
    # Variables maestras
    f = df["f_GHz"].values
    S21 = df["S21"].values
    S21_db = df["S21_db"].values
    S21_mag = df["S21_mag"].values

    # M√©tricas b√°sicas
    n_puntos = len(f)
    f_min, f_max = f[0], f[-1]
    s21_min, s21_max = S21_db.min(), S21_db.max()
    s21_mean = np.mean(S21_db)
    datos_validos = not (np.any(np.isnan(S21_db)) or np.any(np.isinf(S21_db)))

    # Resumen en consola
    print(f"‚úÖ {ARCHIVO_S2P} procesado correctamente")
    print(f"   üìä {n_puntos:,} puntos en el rango {f_min:.3f} ‚Äì {f_max:.3f} GHz")
    print(f"   üìà |S21| en dB: min = {s21_min:.1f}, max = {s21_max:.1f}, promedio = {s21_mean:.1f}")
    print(f"   üîç Integridad de datos: {'‚úì OK' if datos_validos else '‚ö†Ô∏è Problemas detectados'}")

    # Visualizaci√≥n
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

    # Magnitud
    ax1.plot(f, S21_db, color=COLORES["datos"], lw=1.2)
    ax1.set_ylabel("|S21| [dB]")
    ax1.set_title("Magnitud de S21 en funci√≥n de la frecuencia")
    ax1.grid(True, alpha=0.3)

    # Fase
    fase_deg = np.angle(S21, deg=True)
    ax2.plot(f, fase_deg, color=COLORES["comparacion"], lw=1.2)
    ax2.set_ylabel("Fase [¬∞]")
    ax2.set_xlabel("Frecuencia [GHz]")
    ax2.set_title("Fase de S21 en funci√≥n de la frecuencia")
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
else:
    print("‚ùå No se pudieron cargar los datos.")
```

```python
# ============================================================
# üìä M√âTODO SIMPLE (f_res y Q por -3 dB)
# ============================================================

def metodo_simple(f, S21_db, altura_minima=-40, separacion_minima=30, prominencia_min=3, plot=True):
    """
    M√©todo simple para determinar frecuencia de resonancia y Q por -3 dB.

    Par√°metros
    ----------
    f : array
        Frecuencias en GHz
    S21_db : array
        Magnitud de S21 en dB
    altura_minima : float
        Nivel m√≠nimo para considerar un pico (dB)
    separacion_minima : int
        Separaci√≥n m√≠nima entre resonancias (en puntos)
    prominencia_min : float
        Prominencia m√≠nima (criterio interno de find_peaks)
    plot : bool
        Si True, genera gr√°fico con las resonancias detectadas

    Returns
    -------
    df_res : DataFrame con resonancias y Q:
        - numero
        - f_res_GHz
        - S21_res_dB
        - f_izq_3dB
        - f_der_3dB
        - BW_3dB
        - Q_est
    """
    # --- detectar resonancias ---
    idx, _ = signal.find_peaks(
        S21_db,
        height=altura_minima,
        prominence=prominencia_min,
        distance=separacion_minima
    )

    resultados = []
    for i, k in enumerate(idx):
        f_res = f[k]
        nivel_max = S21_db[k]
        nivel_3dB = nivel_max - 3

        # buscar cruces -3 dB
        i_left = np.where(S21_db[:k] <= nivel_3dB)[0]
        i_right = np.where(S21_db[k:] <= nivel_3dB)[0]

        f_izq = f[i_left[-1]] if len(i_left) > 0 else np.nan
        f_der = f[k + i_right[0]] if len(i_right) > 0 else np.nan

        BW = f_der - f_izq if (np.isfinite(f_izq) and np.isfinite(f_der)) else np.nan
        Q_est = f_res / BW if BW and BW > 0 else np.nan

        resultados.append([i+1, f_res, nivel_max, f_izq, f_der, BW, Q_est])

    df_res = pd.DataFrame(resultados, columns=[
        "numero", "f_res_GHz", "S21_res_dB", "f_izq_3dB", "f_der_3dB", "BW_3dB", "Q_est"
    ])

    # --- salida en consola ---
    print("üìä M√âTODO SIMPLE (‚àí3 dB): RESULTADOS")
    print("=" * 60)
    if len(df_res) > 0:
        for _, r in df_res.iterrows():
            print(f"üî∏ Resonancia #{int(r['numero'])}")
            print(f"   ‚îú‚îÄ Frecuencia: {r['f_res_GHz']:.4f} GHz")
            print(f"   ‚îú‚îÄ Magnitud: {r['S21_res_dB']:.2f} dB")
            print(f"   ‚îú‚îÄ f(-3dB) izq: {r['f_izq_3dB']:.4f} GHz")
            print(f"   ‚îú‚îÄ f(-3dB) der: {r['f_der_3dB']:.4f} GHz")
            print(f"   ‚îú‚îÄ Œîf (-3dB): {r['BW_3dB']:.6f} GHz")
            print(f"   ‚îî‚îÄ ‚ú® Q estimado: {r['Q_est']:.1f}\n")
    else:
        print("‚ö†Ô∏è No se detectaron resonancias con los criterios actuales")

    print(f"‚úÖ An√°lisis completado para {len(df_res)} resonancias")

    # --- gr√°fico opcional ---
    if plot and len(df_res) > 0:
        plt.figure(figsize=(12, 6))
        plt.plot(f, S21_db, color=COLORES["datos"], lw=1.2, label="|S21| datos")
        plt.scatter(df_res["f_res_GHz"], df_res["S21_res_dB"],
                    color=COLORES["resonancia"], s=80, marker="o", label="Resonancias")
        for _, row in df_res.iterrows():
            plt.text(row["f_res_GHz"], row["S21_res_dB"]+1,
                     f"{row['f_res_GHz']:.3f} GHz", ha="center", va="bottom",
                     fontsize=9, color=COLORES["resonancia"])
        plt.xlabel("Frecuencia [GHz]")
        plt.ylabel("|S21| [dB]")
        plt.title("M√©todo Simple: detecci√≥n de f_res y Q")
        plt.grid(True, alpha=0.3)
        plt.legend()
        plt.show()

    return df_res
```

```python
# ============================================================
# EJECUCI√ìN DEL M√âTODO SIMPLE Y GUARDADO DE RESULTADOS
# ============================================================
df_simple = metodo_simple(
    f, S21_db,
    altura_minima=-40,
    separacion_minima=30,
    prominencia_min=3,
    plot=True
)
```

```python
# ============================================================
# üìä VISUALIZACI√ìN DETALLADA DE RESULTADOS M√âTODO SIMPLE (-3 dB)
# ============================================================

if df_simple is not None and len(df_simple) > 0:
    n_resonancias = len(df_simple)

    # Layout sim√©trico (ej: 2x2 si hay 4 resonancias)
    if n_resonancias == 4:
        n_rows, n_cols = 2, 2
    elif n_resonancias <= 3:
        n_rows, n_cols = 1, n_resonancias
    else:
        n_cols = 3
        n_rows = -(-n_resonancias // n_cols)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = np.array(axes).reshape(-1)

    for i, row in df_simple.iterrows():
        ax = axes[i]

        # Ventana centrada en la resonancia (8√óBW; fallback 0.05 GHz)
        f_centro = row["f_res_GHz"]
        ancho_ventana = row["BW_3dB"] * 8 if pd.notna(row["BW_3dB"]) else 0.05
        f_min_v, f_max_v = f_centro - ancho_ventana, f_centro + ancho_ventana

        mask = (f >= f_min_v) & (f <= f_max_v)
        f_win, S21_win = f[mask], S21_db[mask]

        ax.plot(f_win, S21_win, color=COLORES["datos"], lw=1.5, label="|S21| medido")
        ax.axvline(x=row["f_res_GHz"], color=COLORES["resonancia"], lw=2, label="f_res")

        nivel_3dB = row["S21_res_dB"] - 3
        ax.axhline(y=nivel_3dB, color=COLORES["destacado"], ls="--", lw=1.5, label="Nivel -3 dB")

        if pd.notna(row["f_izq_3dB"]):
            ax.axvline(x=row["f_izq_3dB"], color=COLORES["destacado"], ls=":", lw=1.5, label="f1 (-3 dB)")
        if pd.notna(row["f_der_3dB"]):
            ax.axvline(x=row["f_der_3dB"], color=COLORES["destacado"], ls=":", lw=1.5, label="f2 (-3 dB)")

        ax.set_xlabel("Frecuencia [GHz]", fontweight="bold")
        ax.set_ylabel("|S21| [dB]", fontweight="bold")
        ax.set_title(f"Resonancia #{int(row['numero'])} ‚Äì Q ‚âà {row['Q_est']:.1f}",
                     fontweight="bold", color=COLORES["ajuste"])
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=9)

    # Ocultar ejes sobrantes
    for j in range(len(df_simple), len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.show()

    print(f"üìä Visualizaci√≥n Q Simple completada para {len(df_simple)} resonancias")
else:
    print("‚ö†Ô∏è No hay resultados Q Simple para visualizar")
```

```python
# ============================================================
# üìê MODELOS LORENTZIANOS
# ============================================================

def modelo_lorentziano_lineal(f, f0, Q, A_real, A_imag, B_real, B_imag):
    A = A_real + 1j * A_imag
    B = B_real + 1j * B_imag
    return A + B / (1 + 2j * Q * (f - f0) / f0)

def modelo_lorentziano_exacto(f, f0, Q, A_real, A_imag, B_real, B_imag):
    A = A_real + 1j * A_imag
    B = B_real + 1j * B_imag
    return A + B / (1 + 2j * Q * ((f/f0) - (f0/f)))
```

```python
def ajustar_lorentziano(f, S21, f0_inicial, Q_inicial=50,
                        ventana=0.15, modo="lineal"):
    """
    Ajusta una resonancia con modelo Lorentziano (lineal o exacto).
    """
    # --- Selecci√≥n de ventana ---
    mask = (f >= f0_inicial - ventana/2) & (f <= f0_inicial + ventana/2)
    f_win = f[mask]
    S21_win = S21[mask]

    if len(f_win) < 10:
        return {"exito": False, "error": "Pocos puntos en la ventana"}

    # --- Modelo seg√∫n modo ---
    if modo == "lineal":
        modelo_func = modelo_lorentziano_lineal
    elif modo == "exacto":
        modelo_func = modelo_lorentziano_exacto
    else:
        raise ValueError("modo debe ser 'lineal' o 'exacto'")

    modelo = Model(modelo_func)

    # Par√°metros iniciales
    params = modelo.make_params(
        f0=f0_inicial,
        Q=Q_inicial,
        A_real=0.0, A_imag=0.0,
        B_real=np.abs(S21_win).max(),
        B_imag=0.0
    )

    # L√≠mites
    params["f0"].set(min=f0_inicial*0.98, max=f0_inicial*1.02)
    params["Q"].set(min=5, max=5000)

    # Residuales (real + imag)
    datos = np.concatenate([S21_win.real, S21_win.imag])
    def residuales(pars, freqs, datos):
        S21_model = modelo_func(freqs, **pars.valuesdict())
        return datos - np.concatenate([S21_model.real, S21_model.imag])

    # Ajuste
    resultado = lmfit.minimize(residuales, params, args=(f_win, datos))
    if not resultado.success:
        return {"exito": False, "error": "Ajuste no convergi√≥"}

    # Modelo ajustado
    S21_fit = modelo_func(f_win, **resultado.params.valuesdict())

    # M√©tricas
    ss_res = np.sum((datos - np.concatenate([S21_fit.real, S21_fit.imag]))**2)
    ss_tot = np.sum((datos - np.mean(datos))**2)
    R2 = 1 - ss_res/ss_tot if ss_tot > 0 else 0

    return {
        "exito": True,
        "f0": resultado.params["f0"].value,
        "Q": resultado.params["Q"].value,
        "params": resultado.params,
        "frecuencias": f_win,
        "S21_datos": S21_win,
        "S21_fit": S21_fit,
        "R2": R2,
        "chi2": resultado.chisqr,
        "chi2_red": resultado.redchi,
        "modo": modo
    }
```

```python
def analizar_lorentziano(df_simple, f, S21, ventana=0.15, modo="lineal"):
    """
    Analiza todas las resonancias con modelo Lorentziano (lineal o exacto).
    """
    resultados = []
    print("="*70)
    print(f"üéØ AN√ÅLISIS LORENTZIANO ({modo.upper()})")
    print("="*70)

    for _, row in df_simple.iterrows():
        numero = int(row["numero"])
        f0_ini = row["f_res_GHz"]
        Q_ini = row["Q_est"]

        print(f"\nüî∏ PROCESANDO RESONANCIA #{numero} ({f0_ini:.4f} GHz)")

        ajuste = ajustar_lorentziano(
            f, S21,
            f0_inicial=f0_ini,
            Q_inicial=Q_ini if not np.isnan(Q_ini) else 50,
            ventana=ventana,
            modo=modo
        )

        if ajuste["exito"]:
            resultados.append(ajuste)
            p = ajuste["params"]
            print("   ‚úÖ AJUSTE EXITOSO:")
            print(f"      ‚îú‚îÄ f‚ÇÄ: {ajuste['f0']:.6f} ¬± {p['f0'].stderr or 0:.6f} GHz")
            print(f"      ‚îú‚îÄ Q: {ajuste['Q']:.1f} ¬± {p['Q'].stderr or 0:.1f}")
            print(f"      ‚îú‚îÄ R¬≤: {ajuste['R2']:.4f}")
            print(f"      ‚îú‚îÄ œá¬≤: {ajuste['chi2']:.2e}")
            print(f"      ‚îî‚îÄ Modo: {modo}")
        else:
            print(f"   ‚ùå AJUSTE FALLIDO: {ajuste['error']}")
    print("\n")
    return resultados
```

```python
# ============================================================
# Ajuste lineal
# ============================================================
resultados_lineal = analizar_lorentziano(
    df_simple, 
    f, 
    S21, 
    ventana=0.8, 
    modo="lineal"
)
# ============================================================
# Ajuste exacto
resultados_exacto = analizar_lorentziano(
    df_simple, 
    f, 
    S21, 
    ventana=0.8, 
    modo="exacto"
)
# ============================================================
```

```python
# ============================================================
# üìä VISUALIZADOR GEN√âRICO - LORENTZIANO (LINEAL O EXACTO)
# ============================================================

def visualizar_lorentziano(resultados):
    """
    Visualiza los resultados de un ajuste Lorentziano (lineal o exacto).
    Muestra magnitud (dB) y fase (grados) para cada resonancia.
    
    Par√°metros
    ----------
    resultados : list of dict
        Salida de analizar_lorentziano (contiene 'modo')
    """
    if len(resultados) == 0:
        print("‚ö†Ô∏è No hay resultados para visualizar")
        return

    modo = resultados[0].get("modo", "desconocido").upper()
    n_res = len(resultados)
    n_cols = 2  # magnitud y fase
    n_rows = n_res

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 4*n_rows))
    fig.suptitle(f"M√©todo Lorentziano ({modo})", fontsize=16, fontweight="bold")

    if n_res == 1:
        axes = np.array([axes])  # caso especial

    for i, res in enumerate(resultados):
        f_win = res["frecuencias"]
        S21_datos = res["S21_datos"]
        S21_fit = res["S21_fit"]

        # Magnitud en dB
        mag_datos = 20*np.log10(np.abs(S21_datos))
        mag_fit = 20*np.log10(np.abs(S21_fit))

        # Fase en grados (unwrap para evitar saltos de 360¬∞)
        # fase_datos = np.unwrap(np.angle(S21_datos)) * 180/np.pi
        # fase_fit = np.unwrap(np.angle(S21_fit)) * 180/np.pi

        fase_datos = np.unwrap(np.angle(S21_datos)) * 180/np.pi
        fase_fit = np.unwrap(np.angle(S21_fit)) * 180/np.pi

        ax_mag = axes[i, 0] if n_res > 1 else axes[0]
        ax_fase = axes[i, 1] if n_res > 1 else axes[1]

        # === Magnitud ===
        ax_mag.plot(f_win, mag_datos, 'o', color=COLORES["datos"],
                    markersize=4, alpha=0.7, label="Datos")
        ax_mag.plot(f_win, mag_fit, '-', color=COLORES["ajuste"],
                    linewidth=2, label=f"Ajuste {modo}")
        ax_mag.axvline(x=res["f0"], color="gray", ls="--", lw=1)
        ax_mag.set_ylabel("|S21| [dB]", fontweight="bold")
        ax_mag.set_title(f"Resonancia {i+1} - Q={res['Q']:.1f}",
                         fontweight="bold", color=COLORES["ajuste"])
        ax_mag.grid(True, alpha=0.3)
        ax_mag.legend(fontsize=9)

        # === Fase ===
        ax_fase.plot(f_win, fase_datos, 'o', color=COLORES["datos"],
                     markersize=4, alpha=0.7, label="Datos")
        ax_fase.plot(f_win, fase_fit, '-', color=COLORES["ajuste"],
                     linewidth=2, label=f"Ajuste {modo}")
        ax_fase.axvline(x=res["f0"], color="gray", ls="--", lw=1)
        ax_fase.set_ylabel("Fase [¬∞]", fontweight="bold")
        ax_fase.set_xlabel("Frecuencia [GHz]", fontweight="bold")
        ax_fase.grid(True, alpha=0.3)
        ax_fase.legend(fontsize=9)

    plt.tight_layout()
    plt.show()

    print(f"üìä Visualizaci√≥n Lorentziano ({modo}) completada para {n_res} resonancias")
```

```python
# ============================================================
# üìä RESUMEN COMPARATIVO ENTRE M√âTODOS (Simple, Lineal, Exacto)
# ============================================================

def resumen_comparativo(df_simple, resultados_lineal, resultados_exacto):
    """
    Genera un resumen comparativo entre los tres m√©todos:
    - M√©todo Simple (-3 dB)
    - Lorentziano Lineal
    - Lorentziano Exacto
    """
    print("üìä RESUMEN COMPARATIVO FINAL - TODAS LAS IMPLEMENTACIONES")
    print("="*90)

    # === Estad√≠sticas globales ===
    resumen = []

    if df_simple is not None and len(df_simple) > 0:
        resumen.append(("M√©todo Simple", "-3 dB", 
                        np.nanmean(df_simple["Q_est"]), np.nanstd(df_simple["Q_est"]), 
                        np.nan, np.nan))

    if len(resultados_lineal) > 0:
        resumen.append(("Lorentziano Lineal", "(f-f‚ÇÄ)/f‚ÇÄ", 
                        np.mean([r["Q"] for r in resultados_lineal]),
                        np.std([r["Q"] for r in resultados_lineal]),
                        np.mean([r["R2"] for r in resultados_lineal]),
                        np.min([r["R2"] for r in resultados_lineal])))

    if len(resultados_exacto) > 0:
        resumen.append(("Lorentziano Exacto", "(f/f‚ÇÄ - f‚ÇÄ/f)", 
                        np.mean([r["Q"] for r in resultados_exacto]),
                        np.std([r["Q"] for r in resultados_exacto]),
                        np.mean([r["R2"] for r in resultados_exacto]),
                        np.min([r["R2"] for r in resultados_exacto])))

    print(f"{'M√©todo':<20} | {'Forma':<15} | {'Q Promedio':<15} | {'R¬≤ Promedio':<15}")
    print("-"*75)

    for metodo, forma, Qmean, Qstd, R2mean, R2min in resumen:
        q_str = f"{Qmean:.1f} ¬± {Qstd:.1f}"
        r2_str = f"{R2mean:.6f}" if not np.isnan(R2mean) else "N/A"
        print(f"{metodo:<20} | {forma:<15} | {q_str:<15} | {r2_str:<15}")

    # === Comparaci√≥n por resonancia ===
    print("\nüîÑ COMPARACI√ìN POR RESONANCIA:")
    print("-"*90)

    n_res = len(df_simple) if df_simple is not None else 0

    for i in range(n_res):
        print(f"\nüìç RESONANCIA #{i+1}:")
        print(f"{'M√©todo':<20} | {'Q':<10} | {'f‚ÇÄ (GHz)':<12}")
        print("-"*45)

        # Simple
        if df_simple is not None and i < len(df_simple):
            row = df_simple.iloc[i]
            print(f"{'M√©todo Simple':<20} | {row['Q_est']:<10.1f} | {row['f_res_GHz']:<12.6f}")

        # Lineal
        if i < len(resultados_lineal):
            r = resultados_lineal[i]
            print(f"{'Lorentziano Lineal':<20} | {r['Q']:<10.1f} | {r['f0']:<12.6f}")

        # Exacto
        if i < len(resultados_exacto):
            r = resultados_exacto[i]
            print(f"{'Lorentziano Exacto':<20} | {r['Q']:<10.1f} | {r['f0']:<12.6f}")

resumen_comparativo(df_simple, resultados_lineal, resultados_exacto)
```

```python
from lmfit import Parameters, minimize

# ===========================================
# Construcci√≥n de ventana GLOBAL para multimodo
# ===========================================
def ventana_multimodo_global(f, f0_list, extra=0.15):
    """
    Recorta un rango global desde min(f0)-extra hasta max(f0)+extra (GHz).
    """
    f0_min, f0_max = np.min(f0_list), np.max(f0_list)
    f1, f2 = f0_min - extra/2, f0_max + extra/2
    f1 = max(f1, f[0]); f2 = min(f2, f[-1])
    mask = (f >= f1) & (f <= f2)
    return f[mask], mask

# ===========================================
# Par√°metros iniciales para multimodo
# ===========================================
def params_iniciales_multimodo(df_simple, f, S21, modo="lineal",
                               usar_slope=False, fref=None, mini_win=0.05):
    """
    Crea un lmfit.Parameters con:
      - A_real, A_imag (offset complejo global)
      - (opcional) C_real, C_imag (pendiente compleja global)
      - Por cada modo k: f0_k, Q_k, Bk_real, Bk_imag
    """
    params = Parameters()
    # Offset global
    params.add("A_real", value=0.0, min=-2, max=2)
    params.add("A_imag", value=0.0, min=-2, max=2)

    # Pendiente compleja opcional: A + C*(f - fref)
    if usar_slope:
        if fref is None:
            fref = np.median(f)
        params.add("C_real", value=0.0, min=-2, max=2)
        params.add("C_imag", value=0.0, min=-2, max=2)
        params.add("fref", value=fref, vary=False)
    else:
        params.add("C_real", value=0.0, vary=False)
        params.add("C_imag", value=0.0, vary=False)
        params.add("fref", value=np.median(f), vary=False)

    # Por-modo
    for i, row in df_simple.iterrows():
        k = int(row["numero"])
        f0k = float(row["f_res_GHz"])
        Qk  = float(row["Q_est"]) if np.isfinite(row["Q_est"]) and row["Q_est"]>0 else 50.0

        # Estimar Bk a partir de una mini ventana alrededor de f0k
        mask_loc = (f >= f0k - mini_win/2) & (f <= f0k + mini_win/2)
        if np.any(mask_loc):
            idx = np.argmax(np.abs(S21[mask_loc]))
            B0 = S21[mask_loc][idx]
        else:
            B0 = 0.1 + 0.0j

        params.add(f"f0_{k}", value=f0k, min=f0k*0.98, max=f0k*1.02)
        params.add(f"Q_{k}", value=Qk,  min=5, max=5000)
        params.add(f"B_real_{k}", value=np.real(B0), min=-10, max=10)
        params.add(f"B_imag_{k}", value=np.imag(B0), min=-10, max=10)

    return params

# ===========================================
# Evaluaci√≥n del modelo multimodo
# ===========================================
def s21_multimodo(f, params, modo="lineal"):
    """
    Devuelve S21 complejo = A + [C*(f - fref)] + sum_k Bk/den_k
    """
    A = params["A_real"].value + 1j*params["A_imag"].value
    C = params["C_real"].value + 1j*params["C_imag"].value
    fref = params["fref"].value

    S = A + C*(f - fref)

    # Suma de modos
    for name in params.keys():
        # buscamos tripletes f0_k, Q_k, B_real_k, B_imag_k
        if name.startswith("f0_"):
            k = name.split("_")[1]
            f0 = params[f"f0_{k}"].value
            Q  = params[f"Q_{k}"].value
            B  = params[f"B_real_{k}"].value + 1j*params[f"B_imag_{k}"].value

            if modo == "lineal":
                den = 1 + 2j*Q*(f - f0)/f0
            elif modo == "exacto":
                den = 1 + 2j*Q*((f/f0) - (f0/f))
            else:
                raise ValueError("modo debe ser 'lineal' o 'exacto'")

            S = S + B/den
    return S

# ===========================================
# Residuales (Real+Imag) y ajuste
# ===========================================
def residuales_multimodo(pars, f, datos, modo="lineal"):
    mod = s21_multimodo(f, pars, modo=modo)
    mod_vec = np.concatenate([mod.real, mod.imag])
    return datos - mod_vec

def ajustar_multimodo(df_simple, f, S21, modo="lineal",
                      ventana_global_abs=0.5, usar_slope=False):
    """
    Ajuste global multimodo (suma de modos) en una ventana global fija.
    - modo: "lineal" o "exacto"
    - ventana_global_abs: ancho total (GHz) extra que envuelve todas las f0
    - usar_slope: a√±ade t√©rmino de l√≠nea base con pendiente compleja
    """
    if df_simple is None or len(df_simple)==0:
        return {"exito": False, "error": "df_simple vac√≠o"}

    f0_list = df_simple["f_res_GHz"].values
    f_win, mask = ventana_multimodo_global(f, f0_list, extra=ventana_global_abs)
    S21_win = S21[mask]

    if len(f_win) < 50:
        return {"exito": False, "error": "Muy pocos puntos en ventana global"}

    # Iniciales
    params = params_iniciales_multimodo(df_simple, f_win, S21_win,
                                        modo=modo, usar_slope=usar_slope)

    # Datos (Real+Imag)
    datos = np.concatenate([S21_win.real, S21_win.imag])

    # Minimizaci√≥n
    out = minimize(residuales_multimodo, params, args=(f_win, datos), kws={"modo": modo})
    if not out.success:
        return {"exito": False, "error": "Ajuste no convergi√≥", "resultado": out}

    # Modelo ajustado y m√©tricas
    S_fit = s21_multimodo(f_win, out.params, modo=modo)
    ss_res = np.sum((datos - np.concatenate([S_fit.real, S_fit.imag]))**2)
    ss_tot = np.sum((datos - np.mean(datos))**2)
    R2 = 1 - ss_res/ss_tot if ss_tot > 0 else 0

    # Extraer tabla de modos
    filas = []
    for _, row in df_simple.iterrows():
        k = int(row["numero"])
        f0 = out.params[f"f0_{k}"].value
        Q  = out.params[f"Q_{k}"].value
        Bc = out.params[f"B_real_{k}"].value + 1j*out.params[f"B_imag_{k}"].value
        filas.append({
            "numero": k,
            "f0_fit_GHz": f0,
            "Q_fit": Q,
            "B_abs": np.abs(Bc),
            "B_phase_deg": np.degrees(np.angle(Bc))
        })
    df_mods = pd.DataFrame(filas).sort_values("numero").reset_index(drop=True)

    return {
        "exito": True,
        "modo": modo,
        "params": out.params,
        "resultado_lmfit": out,
        "frecuencias": f_win,
        "S21_datos": S21_win,
        "S21_fit": S_fit,
        "R2": R2,
        "chi2": out.chisqr,
        "chi2_red": out.redchi,
        "n_puntos": len(f_win),
        "df_modos": df_mods
    }
```

```python
# Multimodo APROXIMADO (lineal)
multi_lineal = ajustar_multimodo(
    df_simple, f, S21,
    modo="lineal",
    ventana_global_abs=0.40,   # ancho total extra que envuelve todas las f0 (aj√∫stalo a gusto)
    usar_slope=False           # activa True si ves baseline inclinada
)

# Multimodo EXACTO
multi_exacto = ajustar_multimodo(
    df_simple, f, S21,
    modo="exacto",
    ventana_global_abs=0.40,
    usar_slope=False
)

# Resumen r√°pido en consola
def imprimir_resumen_multimodo(res):
    if not res["exito"]:
        print("‚ùå", res["error"]); return
    print(f"\nüéØ MULTIMODO ({res['modo'].upper()})")
    print(f"   R¬≤={res['R2']:.6f} | œá¬≤·µ£={res['chi2_red']:.3e} | puntos={res['n_puntos']}")
    print(res["df_modos"].to_string(index=False))

imprimir_resumen_multimodo(multi_lineal)
imprimir_resumen_multimodo(multi_exacto)
```
